# Lista de Exercícios — Vetores e Matrizes em Linguagem C

## Objetivos da lista
- praticar declaração, leitura e percorrimento de vetores e matrizes
- reforçar busca, contagem, soma, média, maior e menor
- exercitar inserção, remoção e ordenação em vetores
- aplicar operações importantes em matrizes, como diagonais, soma de linhas/colunas e transposta
- desenvolver raciocínio de estrutura de dados em linguagem C

## Orientações
1. Resolva os exercícios em linguagem C.
2. Sempre que possível, use laços de repetição para percorrer vetores e matrizes.
3. Evite soluções com complexidade desnecessária.
4. Organize o código com boa indentação.
5. Teste com diferentes valores de entrada.

## Estrutura da lista
- Parte 1: Vetores — nível básico
- Parte 2: Vetores — nível intermediário
- Parte 3: Matrizes — nível básico
- Parte 4: Matrizes — nível intermediário
- Parte 5: Desafios integradores


## Parte 1 — Vetores: nível básico

### Exercício 1 — Leitura e exibição de vetor

Crie um programa que leia **8 números inteiros** para um vetor e, ao final, exiba todos os valores lidos, mostrando também o índice de cada posição.

### Exercício 2 — Soma dos elementos

Leia **10 números inteiros** em um vetor e calcule a **soma total** dos elementos.

### Exercício 3 — Média aritmética

Leia **6 valores reais** em um vetor, calcule a **média aritmética** e mostre o resultado com duas casas decimais.

### Exercício 4 — Maior e menor valor

Leia **12 números inteiros** em um vetor e informe:
- o maior valor
- o menor valor
- as posições em que eles aparecem pela primeira vez

### Exercício 5 — Contagem de pares e ímpares

Leia **15 números inteiros** e conte quantos são **pares** e quantos são **ímpares**.

### Exercício 6 — Busca linear

Leia **10 números inteiros** para um vetor. Em seguida, leia mais um valor e informe se ele está presente no vetor. Caso esteja, mostre a posição da **primeira ocorrência**.

### Exercício 7 — Frequência de um valor

Leia **12 números inteiros** para um vetor. Depois, leia um valor `x` e informe **quantas vezes** ele aparece no vetor.

### Exercício 8 — Impressão em ordem inversa

Leia **10 números inteiros** para um vetor e exiba os elementos em **ordem inversa**.

## Parte 2 — Vetores: nível intermediário

### Exercício 9 — Substituição por condição

Leia **10 números inteiros** em um vetor. Depois, substitua todos os valores **negativos** por `0` e exiba o vetor resultante.

### Exercício 10 — Separação por critério

Leia **12 números inteiros** em um vetor. Depois:
- monte um segundo vetor com os valores **pares**
- monte um terceiro vetor com os valores **ímpares**
- mostre os vetores gerados e seus tamanhos lógicos

### Exercício 11 — Inserção com deslocamento

Considere um vetor com **capacidade física 20** e **tamanho lógico 8**. Leia os 8 valores iniciais. Depois, leia:
- um novo valor
- a posição em que ele deve ser inserido

Insira o elemento na posição indicada, **deslocando os demais para a direita**, e exiba o vetor final.

Valide se a posição é válida e se ainda há espaço disponível no vetor.

### Exercício 12 — Remoção com deslocamento

Considere um vetor com **capacidade física 20** e **tamanho lógico 10**. Leia os 10 valores iniciais e, em seguida, leia a posição que deve ser removida.

Remova o elemento dessa posição, **deslocando os elementos à direita para a esquerda**, atualize o tamanho lógico e mostre o vetor final.

### Exercício 13 — Ordenação crescente com Bubble Sort

Leia **10 números inteiros** e ordene o vetor em **ordem crescente** usando o algoritmo **Bubble Sort**. Ao final, mostre o vetor ordenado.

### Exercício 14 — Ordenação decrescente com Bubble Sort

Leia **10 números inteiros** e ordene o vetor em **ordem decrescente** usando o algoritmo **Bubble Sort**.

### Exercício 15 — Verificação de vetor ordenado

Leia **10 números inteiros** e verifique se o vetor está:
- em ordem crescente
- em ordem decrescente
- ou desordenado

## Parte 3 — Matrizes: nível básico

### Exercício 16 — Leitura e impressão de matriz

Crie um programa que leia uma matriz `3 x 3` de números inteiros e depois a exiba no formato matricial.

### Exercício 17 — Soma total da matriz

Leia uma matriz `4 x 4` de números inteiros e calcule a **soma de todos os elementos**.

### Exercício 18 — Maior e menor elemento da matriz

Leia uma matriz `3 x 4` de números inteiros e informe:
- o maior valor
- o menor valor
- a linha e a coluna de cada um deles na primeira ocorrência

### Exercício 19 — Soma da diagonal principal

Leia uma matriz `3 x 3` e calcule a **soma dos elementos da diagonal principal**.

### Exercício 20 — Soma da diagonal secundária

Leia uma matriz `3 x 3` e calcule a **soma dos elementos da diagonal secundária**.

## Parte 4 — Matrizes: nível intermediário

### Exercício 21 — Soma de cada linha

Leia uma matriz `4 x 4` de números inteiros e mostre a **soma de cada linha** separadamente.

### Exercício 22 — Soma de cada coluna

Leia uma matriz `4 x 4` de números inteiros e mostre a **soma de cada coluna** separadamente.

### Exercício 23 — Contagem de valores pares

Leia uma matriz `5 x 5` de números inteiros e conte quantos elementos são **pares**.

### Exercício 24 — Transposta de matriz

Leia uma matriz `3 x 3` e gere sua **matriz transposta**. Exiba a matriz original e a transposta.

### Exercício 25 — Verificação de matriz identidade

Leia uma matriz `3 x 3` e verifique se ela é uma **matriz identidade**.

Lembrete:
- os elementos da diagonal principal devem ser `1`
- os demais elementos devem ser `0`

### Exercício 26 — Comparação entre diagonais

Leia uma matriz `3 x 3`, calcule a soma da diagonal principal e da diagonal secundária e informe:
- qual soma é maior
- ou se são iguais

## Parte 5 — Desafios integradores

### Exercício 27 — Frequência dos valores em um vetor

Leia **15 números inteiros** em um vetor e informe, para cada valor que aparecer, quantas vezes ele ocorre.

Observação: faça sem usar estruturas mais avançadas como tabela hash. A proposta é reforçar o raciocínio com vetores e laços.

### Exercício 28 — Intercalação de vetores

Leia dois vetores `A` e `B`, cada um com **5 elementos inteiros**, e crie um vetor `C` com **10 posições**, intercalando os elementos:
`A[0], B[0], A[1], B[1], ...`

Exiba o vetor resultante.

### Exercício 29 — Soma de duas matrizes

Leia duas matrizes `3 x 3` de números inteiros e gere uma terceira matriz com a **soma** dos elementos correspondentes.

### Exercício 30 — Pesquisa em matriz

Leia uma matriz `4 x 4` de números inteiros e depois leia um valor `x`. Informe se ele existe na matriz. Caso exista, mostre a **linha** e a **coluna** da primeira ocorrência encontrada.

### Exercício 31 — Tabuleiro com marcação

Crie uma matriz `5 x 5` representando um tabuleiro. Inicialize todos os elementos com `0`. Depois, permita que o usuário informe uma linha e uma coluna para marcar a posição com `1`. Ao final, exiba a matriz.

Desafio extra: permita marcar várias posições até o usuário desejar parar.

### Exercício 32 — Rotina completa de análise de vetor

Leia um vetor de **12 números inteiros** e faça, no mesmo programa:
- soma dos elementos
- média
- maior valor
- menor valor
- quantidade de pares
- quantidade de ímpares
- busca de um valor informado pelo usuário

Esse exercício reúne várias operações clássicas sobre vetores.

## Proposta de organização para entrega

Para cada exercício, o aluno pode seguir este roteiro:
1. análise do problema
2. declaração das variáveis
3. leitura dos dados
4. processamento
5. saída formatada

## Critérios que podem ser observados pelo professor
- uso correto de vetor e matriz
- controle adequado de índices
- uso correto de laços
- clareza na lógica
- organização do código
- correção da saída
- cuidado com limites e tamanho lógico quando necessário


## Sugestões finais

- Comece pelos exercícios de leitura, soma, média e busca.
- Depois avance para inserção, remoção e ordenação em vetores.
- Em matrizes, domine primeiro leitura, impressão e soma total.
- Só depois avance para diagonais, transposta e verificações especiais.
- Tente resolver primeiro sem consultar a solução pronta.
