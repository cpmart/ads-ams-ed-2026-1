# Lista de Exercícios — Funções em C
## Objetivo
Praticar criação e uso de funções na linguagem C.

## Orientações
1. Cada exercício deve possuir **pelo menos uma função separada da `main()`**.
2. Utilize **parâmetros e retorno de valores** sempre que possível.
3. Evite colocar toda a lógica diretamente dentro da `main()`.


## Exercícios Básicos

### Função de Saudação
Crie uma função chamada `saudacao()` que imprima:

```
Bem-vindo ao estudo de funções em C!
```
Chame essa função na `main()`.

### Função de Soma
Crie uma função:

```c
int soma(int a, int b)
```

Ela deve retornar a soma dos dois números.

No `main()`:

* leia dois números
* chame a função
* mostre o resultado.

### Função para verificar número par
Crie uma função:

```c
int ehPar(int numero)
```

Ela deve retornar:

* `1` se for par
* `0` se for ímpar

### Função para calcular quadrado
Crie uma função:

```c
int quadrado(int n)
```

Que retorne o quadrado de um número.

### Função para mostrar números de 1 até N
Crie uma função:

```c
void mostrarNumeros(int n)
```

Ela deve imprimir todos os números de **1 até N**.

## Exercícios Intermediários

### Função para calcular fatorial
Crie uma função:

```c
int fatorial(int n)
```

Utilize um **laço de repetição**.

### Função que verifica número primo
Crie uma função:

```c
int ehPrimo(int n)
```

Retorne:  

- `1` se for primo
- `0` se não for. 

### Função para calcular média
Crie uma função:

```c
float media(float n1, float n2, float n3)
```

Mostre no `main()` se o aluno está **Aprovado ou Reprovado**.

Critério:

```
media >= 6 → Aprovado
```

### Função que mostra tabuada
Crie uma função:

```c
void tabuada(int n)
```

Ela deve imprimir a tabuada de **1 a 10**.

### Função que conta números pares
Crie uma função:

```c
int contarPares(int inicio, int fim)
```

Ela deve contar quantos números pares existem no intervalo.

## Desafios

### Função que calcula potência

```c
int potencia(int base, int expoente)
```

Calcule usando laço de repetição.

### Função que encontra o maior número

```c
int maior(int a, int b, int c)
```

Retorne o maior valor.

### Função que soma números até N

```c
int somaAteN(int n)
```

Exemplo:

```
Entrada: 5
Resultado: 1+2+3+4+5 = 15
```

### Função que inverte um número

```c
int inverterNumero(int n)
```

### Função que verifica palíndromo

```c
int ehPalindromo(int n)
```

Exemplo:

```
121 → é palíndromo
123 → não é
```

## Desafio Final

Crie um programa com **menu usando switch e funções**:

```
1 - Soma
2 - Par ou Ímpar
3 - Fatorial
4 - Tabuada
5 - Verificar primo
0 - Sair
```

Cada opção deve chamar uma função específica.
