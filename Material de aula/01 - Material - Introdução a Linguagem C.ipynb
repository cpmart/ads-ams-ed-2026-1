# O que é a linguagem C?

A linguagem **C** é uma linguagem de programação de propósito geral, muito utilizada no ensino de lógica, algoritmos, sistemas operacionais, compiladores, sistemas embarcados e softwares de alto desempenho.

Ela é considerada uma linguagem importante porque permite ao programador ter bastante controle sobre o funcionamento do programa, incluindo memória e processamento.


# Aplicações da linguagem C

A linguagem C é usada em:

- desenvolvimento de sistemas operacionais
- compiladores
- drivers
- sistemas embarcados
- softwares de alto desempenho
- ensino de programação e lógica

## Exemplo de uso

Com C, podemos criar programas para:

- realizar cálculos

- manipular memória


# Estrutura básica de um programa em C

Um programa em C possui uma estrutura padrão. O ponto de início da execução é a função `main()`.


## Exemplo básico

```c
#include <stdio.h>

int main() {
    printf("Olá, mundo!\n");
    return 0;
}
```


## Explicação da estrutura

### `#include <stdio.h>`
Essa linha inclui uma biblioteca. Nesse caso, a biblioteca `stdio.h` permite usar comandos de entrada e saída, como `printf` e `scanf`.

### `int main()`
É a função principal do programa. Todo programa em C começa a execução por ela.

### `{ }`
As chaves delimitam o bloco de comandos da função.

### `printf("Olá, mundo!\n");`
Exibe uma mensagem na tela.

### `return 0;`
Indica que o programa terminou corretamente.


# Bibliotecas em C

Bibliotecas são conjuntos de funções prontas que podem ser usadas no programa.


## Biblioteca mais comum no início: `stdio.h`

Permite trabalhar com entrada e saída de dados.

Funções comuns:

- `printf()` → mostrar dados na tela
- `scanf()` → ler dados digitados pelo usuário


## Outras bibliotecas conhecidas

- `stdlib.h` → funções utilitárias gerais
- `math.h` → funções matemáticas
- `string.h` → manipulação de strings


## Exemplo com biblioteca matemática

```c
#include <stdio.h>
#include <math.h>

int main() {
    float numero = 25.0;
    printf("Raiz quadrada: %.2f\n", sqrt(numero));
    return 0;
}
```


# Tipos de dados em C

Os tipos de dados definem **que tipo de valor** uma variável pode armazenar.


## `int`

Armazena números inteiros.

```c
int idade = 20;
```

**Aplicações:**

- idade
- quantidade
- número de alunos
- contador


## `float`

Armazena números reais com casas decimais, com precisão simples.

```c
float altura = 1.75;
```

**Aplicações:**

- altura
- peso
- temperatura
- médias


## `double`

Armazena números reais com maior precisão que `float`.

```c
double salario = 2500.75;
```

**Aplicações:**

- cálculos mais precisos
- valores científicos

## `char`

Armazena um único caractere.

```c
char letra = 'A';
```

**Aplicações:**

- letras
- símbolos
- opção de menu
- sexo (`'M'` ou `'F'`)


## Resumo dos tipos

- `int` → número inteiro
- `float` → número real com precisão simples
- `double` → número real com maior precisão
- `char` → um único caractere

# Entrada e saída de dados


## Saída de dados com `printf`

A função `printf` exibe informações na tela.

```c
#include <stdio.h>

int main() {
    int idade = 20;
    printf("Minha idade é %d\n", idade);
    return 0;
}
```


## Especificadores de formato no `printf`

- `%d` → `int`
- `%f` → `float`
- `%lf` → `double`
- `%c` → `char`


## Exemplo com vários tipos

```c
#include <stdio.h>

int main() {
    int idade = 19;
    float altura = 1.80;
    char letra = 'C';

    printf("Idade: %d\n", idade);
    printf("Altura: %.2f\n", altura);
    printf("Letra: %c\n", letra);

    return 0;
}
```

O `%.2f` indica que o número real será exibido com 2 casas decimais.


## Entrada de dados com `scanf`

A função `scanf` permite ler dados digitados pelo usuário.

```c
#include <stdio.h>

int main() {
    int idade;
    printf("Digite sua idade: ");
    scanf("%d", &idade);

    printf("Você digitou: %d\n", idade);
    return 0;
}
```


## Importância do `&`

No `scanf`, geralmente usamos `&` antes da variável para indicar o endereço de memória onde o valor será armazenado.

Exemplo:

```c
scanf("%d", &idade);
```


## Exemplo com `float`

```c
#include <stdio.h>

int main() {
    float nota;
    printf("Digite a nota: ");
    scanf("%f", &nota);

    printf("Nota informada: %.2f\n", nota);
    return 0;
}
```


## Exemplo com `char`

```c
#include <stdio.h>

int main() {
    char sexo;
    printf("Digite M ou F: ");
    scanf(" %c", &sexo);

    printf("Valor digitado: %c\n", sexo);
    return 0;
}
```

Observe o espaço antes de `%c` em `scanf(" %c", &sexo);`.  
Esse espaço ajuda a ignorar o `Enter` que ficou no buffer do teclado.


# Operadores em C

Operadores são símbolos usados para realizar operações com valores e variáveis.


## Operadores aritméticos

Usados em cálculos matemáticos.

- `+` → adição
- `-` → subtração
- `*` → multiplicação
- `/` → divisão
- `%` → resto da divisão

```c
#include <stdio.h>

int main() {
    int a = 10, b = 3;

    printf("Soma: %d\n", a + b);
    printf("Subtração: %d\n", a - b);
    printf("Multiplicação: %d\n", a * b);
    printf("Divisão inteira: %d\n", a / b);
    printf("Resto: %d\n", a % b);

    return 0;
}
```


## Observação importante sobre divisão

Quando dois inteiros são divididos em C, o resultado é inteiro.

Exemplo:

```c
10 / 3
```

Resultado:

```c
3
```

Se quiser divisão real:

```c
float resultado = 10.0 / 3.0;
```


## Operadores relacionais

Usados para comparar valores. O resultado será verdadeiro ou falso.

- `==` → igual a
- `!=` → diferente de
- `>` → maior que
- `<` → menor que
- `>=` → maior ou igual a
- `<=` → menor ou igual a

```c
#include <stdio.h>

int main() {
    int a = 10, b = 5;

    printf("%d\n", a > b);
    printf("%d\n", a == b);
    printf("%d\n", a != b);

    return 0;
}
```

Em C, normalmente:

- `1` representa verdadeiro
- `0` representa falso


## Operadores lógicos

Usados para combinar condições.

- `&&` → E
- `||` → OU
- `!` → NÃO

```c
#include <stdio.h>

int main() {
    int idade = 20;
    int possui_carteira = 1;

    printf("%d\n", idade >= 18 && possui_carteira == 1);

    return 0;
}
```


## Operadores de atribuição

Usados para atribuir valores às variáveis.

- `=` → atribuição simples
- `+=` → soma e atribui
- `-=` → subtrai e atribui
- `*=` → multiplica e atribui
- `/=` → divide e atribui

```c
#include <stdio.h>

int main() {
    int x = 10;
    x += 5;
    printf("%d\n", x);

    return 0;
}
```


## Operadores de incremento e decremento

- `++` → incrementa 1
- `--` → decrementa 1

```c
#include <stdio.h>

int main() {
    int x = 5;

    x++;
    printf("%d\n", x);

    x--;
    printf("%d\n", x);

    return 0;
}
```


# Exemplo completo integrando os conceitos

```c
#include <stdio.h>

int main() {
    int idade;
    float altura;
    char inicial;

    printf("Digite sua idade: ");
    scanf("%d", &idade);

    printf("Digite sua altura: ");
    scanf("%f", &altura);

    printf("Digite a inicial do seu nome: ");
    scanf(" %c", &inicial);

    printf("\n--- Dados informados ---\n");
    printf("Idade: %d\n", idade);
    printf("Altura: %.2f\n", altura);
    printf("Inicial: %c\n", inicial);

    return 0;
}
```


## O que esse programa faz?

- declara variáveis de tipos diferentes
- lê dados do usuário
- exibe os dados na tela
- usa a biblioteca `stdio.h`
- utiliza entrada e saída de dados


# Exemplos práticos de aplicação


## Exemplo: Soma de dois números

```c
#include <stdio.h>

int main() {
    int a, b, soma;

    printf("Digite o primeiro número: ");
    scanf("%d", &a);

    printf("Digite o segundo número: ");
    scanf("%d", &b);

    soma = a + b;

    printf("Soma = %d\n", soma);

    return 0;
}
```


## Exemplo: Cálculo de média

```c
#include <stdio.h>

int main() {
    float n1, n2, media;

    printf("Digite a primeira nota: ");
    scanf("%f", &n1);

    printf("Digite a segunda nota: ");
    scanf("%f", &n2);

    media = (n1 + n2) / 2;

    printf("Média = %.2f\n", media);

    return 0;
}
```


## Exemplo: Verificar resto da divisão

```c
#include <stdio.h>

int main() {
    int numero;

    printf("Digite um número inteiro: ");
    scanf("%d", &numero);

    printf("Resto da divisão por 2: %d\n", numero % 2);

    return 0;
}
```


# Erros comuns de iniciantes


## Esquecer ponto e vírgula

Errado:

```c
int x = 10
```

Certo:

```c
int x = 10;
```


## Usar especificador errado no `printf` ou `scanf`

Errado:

```c
float nota;
scanf("%d", &nota);
```

Certo:

```c
float nota;
scanf("%f", &nota);
```


## Esquecer o `&` no `scanf`

Errado:

```c
scanf("%d", idade);
```

Certo:

```c
scanf("%d", &idade);
```


## Confundir `=` com `==`

- `=` → atribuição
- `==` → comparação

Exemplo de atribuição:

```c
x = 5;
```

Exemplo de comparação:

```c
x == 5
```
