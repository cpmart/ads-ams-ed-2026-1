# O que é uma estrutura condicional?

Estruturas condicionais são comandos usados para permitir que um programa **tome decisões**.

Elas fazem o programa verificar uma condição e, a partir disso, escolher qual bloco de instruções deverá ser executado.

Em termos simples:

- **se** uma condição for verdadeira, o programa faz uma ação;
- **senão**, ele pode fazer outra ação.

As estruturas condicionais são fundamentais porque tornam os programas mais inteligentes, dinâmicos e adaptáveis.

# Aplicações das estruturas condicionais

As estruturas condicionais podem ser usadas em muitas situações do dia a dia da programação, como:

- verificar se um aluno foi aprovado ou reprovado;
- identificar se um número é positivo, negativo ou zero;
- conferir se uma senha está correta;
- validar idade para acesso a determinado conteúdo;
- criar menus de opções;
- decidir se um desconto deve ser aplicado;
- tratar divisões por zero;
- classificar dados de entrada.

Sem estruturas condicionais, o programa executaria sempre a mesma sequência de comandos, sem capacidade de decisão.

# Principais estruturas condicionais em C

Na linguagem C, as principais estruturas condicionais são:

- `if`
- `if ... else`
- `if ... else if ... else`
- `switch`

Cada uma é usada em situações específicas.

# Estrutura `if`

A estrutura `if` executa um bloco de comandos **somente se a condição for verdadeira**.

## Sintaxe

```c
if (condicao) {
    // comandos
}
```

````python
#include <stdio.h>

int main() {
    int idade = 20;

    if (idade >= 18) {
        printf("Voce e maior de idade.\n");
    }

    return 0;
}
````

## Explicação do exemplo

No exemplo acima:

- a variável `idade` recebe `20`;
- o programa verifica se `idade >= 18`;
- como a condição é verdadeira, a mensagem é exibida.

Se a condição fosse falsa, nada seria executado dentro do `if`.

# Estrutura `if ... else`

A estrutura `if ... else` é usada quando queremos definir **dois caminhos possíveis**:

- um para quando a condição é verdadeira;
- outro para quando a condição é falsa.

## Sintaxe

```c
if (condicao) {
    // comandos se verdadeiro
} else {
    // comandos se falso
}
```

````python
#include <stdio.h>

int main() {
    int numero;

    printf("Digite um numero: ");
    scanf("%d", &numero);

    if (numero % 2 == 0) {
        printf("O numero e par.\n");
    } else {
        printf("O numero e impar.\n");
    }

    return 0;
}
````

## Explicação do exemplo

- O operador `%` calcula o resto da divisão.
- Se o resto da divisão por 2 for igual a zero, o número é par.
- Caso contrário, o número é ímpar.

# Estrutura `if ... else if ... else`

Essa estrutura é usada quando existem **várias possibilidades** e queremos testar diferentes condições em sequência.

## Sintaxe

```c
if (condicao1) {
    // bloco 1
} else if (condicao2) {
    // bloco 2
} else {
    // bloco final
}
```

````python
#include <stdio.h>

int main() {
    int numero;

    printf("Digite um numero: ");
    scanf("%d", &numero);

    if (numero > 0) {
        printf("Numero positivo.\n");
    } else if (numero < 0) {
        printf("Numero negativo.\n");
    } else {
        printf("Numero igual a zero.\n");
    }

    return 0;
}
````

## Explicação do exemplo

O programa analisa:

1. se o número é maior que zero;
2. se não for, verifica se é menor que zero;
3. se não for nenhuma das opções anteriores, então ele é igual a zero.

# Estrutura `switch`

A estrutura `switch` é útil quando queremos comparar uma variável com **valores fixos**, geralmente em menus e seleção de opções.

## Sintaxe

```c
switch (expressao) {
    case valor1:
        // comandos
        break;
    case valor2:
        // comandos
        break;
    default:
        // comandos
}
```

````python
#include <stdio.h>

int main() {
    int opcao;

    printf("Escolha uma opcao:\n");
    printf("1 - Cadastrar\n");
    printf("2 - Alterar\n");
    printf("3 - Excluir\n");
    printf("Digite a opcao: ");
    scanf("%d", &opcao);

    switch (opcao) {
        case 1:
            printf("Opcao cadastrar selecionada.\n");
            break;
        case 2:
            printf("Opcao alterar selecionada.\n");
            break;
        case 3:
            printf("Opcao excluir selecionada.\n");
            break;
        default:
            printf("Opcao invalida.\n");
    }

    return 0;
}
````

### Observação importante

O comando `break` é importante para interromper a execução do `switch`.

Sem o `break`, o programa pode continuar executando os próximos `case`.

# Operadores relacionais

As condições normalmente usam operadores relacionais para comparar valores.

- `==` — **igual a**: verifica se dois valores são iguais
- `!=` — **diferente de**: verifica se dois valores são diferentes
- `>` — **maior que**: verifica se o valor da esquerda é maior
- `<` — **menor que**: verifica se o valor da esquerda é menor
- `>=` — **maior ou igual a**: verifica se o valor da esquerda é maior ou igual
- `<=` — **menor ou igual a**: verifica se o valor da esquerda é menor ou igual

## Exemplo

```c
if (nota >= 6) {
    printf("Aprovado\n");
}
```

# Operadores lógicos

Também é possível combinar condições usando operadores lógicos.

- `&&` — **E (AND)**: a condição é verdadeira somente se **ambas** as partes forem verdadeiras
- `||` — **OU (OR)**: a condição é verdadeira se **pelo menos uma** das partes for verdadeira
- `!` — **NÃO (NOT)**: inverte o resultado da condição (verdadeiro vira falso e vice-versa)

````python
#include <stdio.h>

int main() {
    int idade = 25;

    if (idade >= 18 && idade <= 65) {
        printf("Idade dentro da faixa.\n");
    }

    return 0;
}
````

````python
#include <stdio.h>

int main() {
    int dia = 7;

    if (dia == 1 || dia == 7) {
        printf("Fim de semana.\n");
    }

    return 0;
}
````

````python
#include <stdio.h>

int main() {
    int numero = 10;

    if (!(numero == 0)) {
        printf("O numero nao e zero.\n");
    }

    return 0;
}
````

# Exemplos práticos completos

## Exemplo 1: Aprovação do aluno

````python
#include <stdio.h>

int main() {
    float nota;

    printf("Digite a nota do aluno: ");
    scanf("%f", &nota);

    if (nota >= 6.0) {
        printf("Aluno aprovado.\n");
    } else {
        printf("Aluno reprovado.\n");
    }

    return 0;
}
````

## Exemplo 2: Verificar o maior de dois números

````python
#include <stdio.h>

int main() {
    int a, b;

    printf("Digite o primeiro numero: ");
    scanf("%d", &a);

    printf("Digite o segundo numero: ");
    scanf("%d", &b);

    if (a > b) {
        printf("O maior numero e %d\n", a);
    } else if (b > a) {
        printf("O maior numero e %d\n", b);
    } else {
        printf("Os numeros sao iguais.\n");
    }

    return 0;
}
````

## Exemplo 3: Calculadora simples com `switch`

````python
#include <stdio.h>

int main() {
    int op;
    float n1, n2;

    printf("Digite dois numeros: ");
    scanf("%f %f", &n1, &n2);

    printf("1 - Somar\n");
    printf("2 - Subtrair\n");
    printf("3 - Multiplicar\n");
    printf("4 - Dividir\n");
    printf("Escolha: ");
    scanf("%d", &op);

    switch (op) {
        case 1:
            printf("Resultado: %.2f\n", n1 + n2);
            break;
        case 2:
            printf("Resultado: %.2f\n", n1 - n2);
            break;
        case 3:
            printf("Resultado: %.2f\n", n1 * n2);
            break;
        case 4:
            if (n2 != 0) {
                printf("Resultado: %.2f\n", n1 / n2);
            } else {
                printf("Nao e possivel dividir por zero.\n");
            }
            break;
        default:
            printf("Opcao invalida.\n");
    }

    return 0;
}
````

# Erros comuns

## Confundir `=` com `==`

- `=` significa **atribuição**
- `==` significa **comparação**

Errado:

```c
if (x = 5)
```

Certo:

```c
if (x == 5)
```

## Esquecer chaves

Mesmo quando há apenas uma instrução, usar `{}` ajuda na organização e evita erros.

## Escrever condições mal planejadas

É importante pensar na ordem das comparações, principalmente quando existem várias faixas de valores.
