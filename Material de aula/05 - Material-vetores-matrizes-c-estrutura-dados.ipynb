# Ideia central: estrutura homogênea e indexada

Antes de falar em vetor, é importante entender duas ideias fundamentais.

## Estrutura homogênea

Uma estrutura homogênea armazena elementos do **mesmo tipo**.

Exemplo:
```c
int idades[30];
```

## Estrutura indexada

Os elementos são acessados por **posição**.

Em C, a primeira posição é sempre a posição `0`.

Exemplo:
```text
índice:  0   1   2   3   4
valor:   7  11  25   9  18
```

# Vetor: conceito formal e intuição

Um **vetor** é uma estrutura de dados linear, homogênea e indexada, composta por um conjunto finito de elementos armazenados em posições contíguas de memória.

## Intuição

Pense em uma fileira de caixas numeradas:

```text
[ 12 ][ 19 ][  4 ][ 30 ][ 15 ]
   0     1     2     3     4
```

Cada caixa guarda um valor.
Para acessar um valor específico, basta informar a posição.

## Representação em C

```c
int v[5];
```

# Vetor e memória

Em C, os elementos de um vetor ficam em **endereços contíguos de memória**.

Se um `int` ocupa 4 bytes, então um vetor como:

```c
int v[5] = {10, 20, 30, 40, 50};
```

pode ser imaginado assim:

```text
índice      valor      endereço (exemplo)
0           10         1000
1           20         1004
2           30         1008
3           40         1012
4           50         1016
```

A operação:

```c
v[i]
```

acessa diretamente a posição correta, o que explica a rapidez do acesso por índice.

# Busca linear

A busca linear percorre o vetor do início ao fim comparando cada elemento com o valor procurado.

## Algoritmo

1. começar em `i = 0`
2. comparar `v[i]` com o valor desejado
3. se encontrar, parar
4. se chegar ao fim, concluir que não existe

````python
#include <stdio.h>

int main() {
    int v[8] = {12, 7, 30, 18, 9, 21, 5, 14};
    int valor;
    int posicao = -1;

    printf("Digite o valor a buscar: ");
    scanf("%d", &valor);

    for (int i = 0; i < 8; i++) {
        if (v[i] == valor) {
            posicao = i;
            break;
        }
    }

    if (posicao != -1) {
        printf("Valor encontrado na posicao %d\n", posicao);
    } else {
        printf("Valor nao encontrado\n");
    }

    return 0;
}
````

## Custo da busca linear

No pior caso, o algoritmo verifica todas as posições.

Se o vetor tem `n` elementos, o custo é:

```text
O(n)
```

# Inserção e remoção em vetores

## Inserção no meio

Se quiser inserir em uma posição específica, é necessário deslocar elementos para a direita.

Antes:
```text
[10][20][30][40]
```

Depois de inserir `99` na posição 2:
```text
[10][20][99][30][40]
```

````python
#include <stdio.h>

int main() {
    int v[10] = {10, 20, 30, 40, 50};
    int tamanho = 5;
    int posicao = 2;
    int novo = 99;

    for (int i = tamanho; i > posicao; i--) {
        v[i] = v[i - 1];
    }

    v[posicao] = novo;
    tamanho++;

    for (int i = 0; i < tamanho; i++) {
        printf("%d ", v[i]);
    }
    printf("\n");

    return 0;
}
````

## Remoção

Na remoção, os elementos à direita devem ser deslocados para a esquerda.

````python
#include <stdio.h>

int main() {
    int v[10] = {10, 20, 30, 40, 50};
    int tamanho = 5;
    int posicao = 2;

    for (int i = posicao; i < tamanho - 1; i++) {
        v[i] = v[i + 1];
    }

    tamanho--;

    for (int i = 0; i < tamanho; i++) {
        printf("%d ", v[i]);
    }
    printf("\n");

    return 0;
}
````

## Conclusão importante sobre vetores

Vetores têm:
- **acesso rápido por índice**
- **busca linear simples**
- **inserção e remoção no meio custosas**

# Operações clássicas de processamento

Problemas muito comuns:
- soma e média
- maior e menor
- contagem por condição
- frequência de um valor
- impressão inversa

````python
#include <stdio.h>

int main() {
    int v[6] = {8, 3, 15, 2, 10, 7};
    int maior = v[0];
    int menor = v[0];
    int soma = 0;

    for (int i = 0; i < 6; i++) {
        soma += v[i];

        if (v[i] > maior) {
            maior = v[i];
        }

        if (v[i] < menor) {
            menor = v[i];
        }
    }

    printf("Soma = %d\n", soma);
    printf("Media = %.2f\n", soma / 6.0);
    printf("Maior = %d\n", maior);
    printf("Menor = %d\n", menor);

    return 0;
}
````

# Ordenação: primeiro contato

O algoritmo mais didático para início é o **Bubble Sort**.

Ele ajuda a explicar:
- comparação
- troca
- múltiplas passagens pelo vetor
- crescimento de custo

````python
#include <stdio.h>

int main() {
    int v[6] = {5, 1, 4, 2, 8, 3};
    int aux;

    for (int i = 0; i < 5; i++) {
        for (int j = 0; j < 5 - i; j++) {
            if (v[j] > v[j + 1]) {
                aux = v[j];
                v[j] = v[j + 1];
                v[j + 1] = aux;
            }
        }
    }

    for (int i = 0; i < 6; i++) {
        printf("%d ", v[i]);
    }
    printf("\n");

    return 0;
}
````

## Custo do Bubble Sort

No caso geral, o custo aproximado é:

```text
O(n²)
```

# Matriz: ampliando a ideia do vetor

Uma **matriz** é uma estrutura homogênea bidimensional.

Ela usa:
- **linhas**
- **colunas**

## Representação visual

```text
        coluna
         0  1  2
linha 0  5  7  1
linha 1  3  9  4
linha 2  8  6  2
```

Em C:

```c
matriz[linha][coluna]
```

# Declaração e leitura de matrizes

## Declaração

```c
int m[3][3];
```

## Inicialização

```c
int m[3][3] = {
    {1, 2, 3},
    {4, 5, 6},
    {7, 8, 9}
};
```

## Percorrimento

Usamos dois laços:
- um para linha
- outro para coluna

````python
#include <stdio.h>

int main() {
    int m[3][3];

    for (int i = 0; i < 3; i++) {
        for (int j = 0; j < 3; j++) {
            printf("Digite m[%d][%d]: ", i, j);
            scanf("%d", &m[i][j]);
        }
    }

    printf("\nMatriz lida:\n");
    for (int i = 0; i < 3; i++) {
        for (int j = 0; j < 3; j++) {
            printf("%d ", m[i][j]);
        }
        printf("\n");
    }

    return 0;
}
````

# Operações clássicas em matrizes

- leitura e impressão
- soma de todos os elementos
- maior e menor valor
- soma de linhas e colunas
- diagonal principal
- diagonal secundária
- transposta

````python
#include <stdio.h>

int main() {
    int m[3][3] = {
        {2, 4, 6},
        {1, 3, 5},
        {7, 8, 9}
    };

    int soma = 0;
    int diagonal = 0;

    for (int i = 0; i < 3; i++) {
        for (int j = 0; j < 3; j++) {
            soma += m[i][j];

            if (i == j) {
                diagonal += m[i][j];
            }
        }
    }

    printf("Soma total = %d\n", soma);
    printf("Soma da diagonal principal = %d\n", diagonal);

    return 0;
}
````

# Matriz transposta

A transposta de uma matriz é obtida trocando linhas por colunas.

Em termos de índice:

```text
T[j][i] = A[i][j]
```

````python
#include <stdio.h>

int main() {
    int a[3][3] = {
        {1, 2, 3},
        {4, 5, 6},
        {7, 8, 9}
    };

    int t[3][3];

    for (int i = 0; i < 3; i++) {
        for (int j = 0; j < 3; j++) {
            t[j][i] = a[i][j];
        }
    }

    printf("Matriz transposta:\n");
    for (int i = 0; i < 3; i++) {
        for (int j = 0; j < 3; j++) {
            printf("%d ", t[i][j]);
        }
        printf("\n");
    }

    return 0;
}
````

# Complexidade em matrizes

Se um vetor com `n` elementos é percorrido uma vez, o custo é:

```text
O(n)
```

Já em uma matriz `n x n`, um percorrimento completo visita `n²` posições:

```text
O(n²)
```

# Capacidade física x tamanho lógico

## Capacidade física
É a quantidade máxima de posições reservadas.

Exemplo:
```c
int v[100];
```

## Tamanho lógico
É a quantidade de elementos realmente usados naquele momento.

Exemplo:
- capacidade 100
- elementos usados 12
- tamanho lógico 12
