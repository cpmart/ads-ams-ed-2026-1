## O que é uma função?

Uma **função** é um bloco de código que executa uma tarefa específica.

Ela serve para:

- organizar o programa
- evitar repetição de código
- facilitar manutenção
- dividir problemas grandes em partes menores

### Ideia principal

```
Programa grande
      ↓
dividido em
      ↓
funções menores
```


## Exemplo simples

```c
#include <stdio.h>

void saudacao() {
    printf("Bem-vindo ao estudo de funções!\n");
}

int main() {
    saudacao();
    return 0;
}
```

Saída esperada:

```
Bem-vindo ao estudo de funções!
```


## Estrutura de uma função

Uma função em C possui três partes principais:

```c
tipo_retorno nome_da_funcao(parametros) {
    codigo
}
```

### Explicação da estrutura

**Tipo de retorno**
- define qual tipo de valor a função devolve
- exemplos: `int`, `float`, `double`, `void`

**Nome da função**
- identifica a função
- deve representar claramente o que ela faz

**Parâmetros**
- valores que a função recebe para executar sua tarefa

**Return**
- instrução que envia um valor de volta para quem chamou a função


## Onde a função é executada?

Funções são executadas quando são **chamadas**.

```c
#include <stdio.h>

int soma(int a, int b) {
    return a + b;
}

int main() {

    int resultado;

    resultado = soma(5,3);

    printf("Resultado: %d\n", resultado);

    return 0;
}
```

Fluxo de execução:

```
main()
 ↓
chama soma()
 ↓
soma calcula
 ↓
retorna valor
 ↓
main continua
```


## Tipos de funções

Existem quatro tipos principais de funções em C.


### Função sem parâmetro e sem retorno

Executa apenas uma ação.

```c
void mensagem() {
    printf("Aprendendo funções!\n");
}
```

Uso:

```c
mensagem();
```


### Função com parâmetro e sem retorno

Recebe dados, mas apenas executa uma tarefa.

```c
void mostrarNumero(int n) {
    printf("Numero: %d\n", n);
}
```

Uso:

```c
mostrarNumero(10);
```


### Função com parâmetro e com retorno

Muito comum em programas.

```c
int quadrado(int n) {
    return n * n;
}
```

Uso:

```c
int x = quadrado(5);
```


### Função sem parâmetro com retorno

Mais raro, mas possível.

```c
int gerarNumero() {
    return 10;
}
```

Uso:

```c
int valor = gerarNumero();
```


## Exemplo completo

```c
#include <stdio.h>

int soma(int a, int b) {
    return a + b;
}

int main() {

    int n1, n2, resultado;

    printf("Digite dois numeros: ");
    scanf("%d %d", &n1, &n2);

    resultado = soma(n1, n2);

    printf("Soma = %d\n", resultado);

    return 0;
}
```


## Funções com estruturas condicionais

Funções podem usar `if`.

```c
int ehPar(int n) {

    if (n % 2 == 0)
        return 1;
    else
        return 0;
}
```

Uso:

```c
if (ehPar(numero))
    printf("Numero par\n");
else
    printf("Numero impar\n");
```


## Funções com laços de repetição

Também podem conter `for` ou `while`.

### Exemplo: tabuada

```c
void tabuada(int n) {

    for(int i = 1; i <= 10; i++) {
        printf("%d x %d = %d\n", n, i, n*i);
    }

}
```

Uso:

```c
tabuada(7);
```


## Escopo de variáveis

Variáveis criadas dentro de uma função **existem apenas nela**.

```c
void exemplo() {
    int x = 10;
}
```

A variável `x` não pode ser utilizada fora da função.


## Boas práticas ao criar funções

- cada função deve resolver **apenas um problema**
- utilize **nomes claros**
- evite funções muito grandes
- reutilize funções sempre que possível
- prefira programas compostos por várias funções pequenas


## Exemplo ruim

```c
void tudo() {

    // lê número
    // calcula média
    // imprime resultado
    // verifica aprovação
}
```


## Exemplo melhor

Separar o programa em funções menores:

```
lerNumero()
calcularMedia()
verificarAprovacao()
mostrarResultado()
```

Estrutura do programa:

```
main()
 ├── lerNumero()
 ├── calcularMedia()
 ├── verificarAprovacao()
 └── mostrarResultado()
```


## Curiosidade

Bibliotecas em C possuem muitas funções prontas.

Exemplos da biblioteca matemática:

- `sqrt()` → raiz quadrada
- `pow()` → potência
- `sin()` → seno
- `cos()` → cosseno

Exemplo:

```c
#include <math.h>

double r = sqrt(25);
```

Resultado:

```
5
```

```c
soma(int a, int b)
```

```c
int soma(int a, int b)
```


### Não retornar valor

```c
int soma(int a, int b) {
    int r = a + b;
}
```

```c
int soma(int a, int b) {
    return a + b;
}
```

```c
int soma(int a, int b);

int main() {

}
```

## Exercício conceitual

Qual será a saída?

```c
int dobro(int n) {
    return n * 2;
}

int main() {

    int x = 5;

    printf("%d", dobro(x));

}
```

Resposta:

```
10
```


## Exercício de raciocínio

O que será impresso?

```c
int soma(int a, int b) {
    return a + b;
}

int main() {

    int x = soma(3,4);

    printf("%d", soma(x,2));

}
```

Passos:

```
soma(3,4) = 7
soma(7,2) = 9
```

Saída:

```
9
```
