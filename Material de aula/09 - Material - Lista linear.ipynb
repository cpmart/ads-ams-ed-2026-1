# O que é uma lista linear?

Uma **lista linear** é uma estrutura de dados em que os elementos são organizados em uma sequência.

Cada elemento possui uma posição lógica dentro da lista.

Exemplo:

```text
[10] [20] [30] [40] [50]
```

Essa organização é chamada de linear porque existe uma relação de ordem entre os elementos:

```text
10 vem antes de 20
20 vem antes de 30
30 vem antes de 40
40 vem antes de 50
```

Em uma lista linear, normalmente realizamos operações como:

- inserir elementos;
- remover elementos;
- buscar elementos;
- percorrer todos os elementos;
- alterar valores;
- verificar se a lista está vazia;
- verificar o tamanho da lista.

# Formas de implementar uma lista linear

Uma lista linear pode ser implementada de várias formas. Nesta aula, estudaremos as seguintes:

```text
Lista Linear
|
|-- Vetor
|
|-- Lista Encadeada Simples
|
|-- Lista Encadeada Circular
|
|-- Lista Duplamente Encadeada
|
|-- Lista Duplamente Encadeada Circular
```

A principal diferença entre essas estruturas está na forma como os dados são armazenados e ligados na memória.

# Lista linear usando vetor

## Conceito

Um **vetor** é uma sequência de posições consecutivas na memória.

Em C, podemos declarar um vetor de inteiros assim:

```c
int numeros[5];
```

Esse vetor possui 5 posições:

```text
Índice:   0    1    2    3    4
Valor:   ?    ?    ?    ?    ?
```

Depois de preencher o vetor:

```c
int numeros[5] = {10, 20, 30, 40, 50};
```

Temos:

```text
Índice:   0    1    2    3    4
Valor:   10   20   30   40   50
```

## Observação importante

Em C, os índices começam em `0`.

Portanto, em um vetor com 5 elementos, os índices válidos são:

```text
0, 1, 2, 3, 4
```

O índice `5` não pertence ao vetor.

# Características do vetor

O vetor possui algumas características importantes:

```text
1. Os elementos ficam em posições consecutivas da memória.

2. O acesso a qualquer posição é direto.

3. O tamanho normalmente é definido antes do uso.

4. Inserções e remoções no meio exigem deslocamento de elementos.

5. É uma estrutura simples e eficiente quando o tamanho é conhecido.
```

Exemplo de acesso direto:

```c
printf("%d", numeros[2]);
```

Se `numeros[2]` contém `30`, a saída será:

```text
30
```

# Criando e exibindo uma lista usando vetor

Neste exemplo, usamos um vetor com capacidade para 10 elementos, mas inicialmente armazenamos apenas 3 valores.

A variável `quantidade` controla quantos elementos estão sendo usados.

````python
%%writefile codigos_c/exemplo_01.c

#include <stdio.h>

#define TAM 10

int main() {
    int lista[TAM];
    int quantidade = 0;

    lista[0] = 10;
    lista[1] = 20;
    lista[2] = 30;

    quantidade = 3;

    printf("Elementos da lista:\n");

    for (int i = 0; i < quantidade; i++) {
        printf("%d ", lista[i]);
    }

    return 0;
}
````

# Inserção no final do vetor

Inserir no final é simples quando ainda existe espaço disponível.

Situação inicial:

```text
10 20 30
```

Inserindo `40` no final:

```text

```

````python
%%writefile codigos_c/exemplo_02.c

#include <stdio.h>

#define TAM 10

int main() {
    int lista[TAM] = {10, 20, 30};
    int quantidade = 3;

    if (quantidade < TAM) {
        lista[quantidade] = 40;
        quantidade++;
    } else {
        printf("Lista cheia.\n");
    }

    printf("Lista após inserção:\n");

    for (int i = 0; i < quantidade; i++) {
        printf("%d ", lista[i]);
    }

    return 0;
}
````

# Inserção em uma posição específica do vetor

Para inserir um elemento no meio do vetor, precisamos deslocar os elementos para a direita.

Exemplo: inserir `25` na posição `2`.

Antes:

```text
Índice:   0    1    2    3
Valor:   10   20   30   40
```

Depois:

```text
Índice:   0    1    2    3    4
Valor:   10   20   25   30   40
```

O valor `30` e o valor `40` foram deslocados.

````python
%%writefile codigos_c/exemplo_03.c

#include <stdio.h>

#define TAM 10

int main() {
    int lista[TAM] = {10, 20, 30, 40};
    int quantidade = 4;
    int valor = 25;
    int posicao = 2;

    if (quantidade < TAM && posicao >= 0 && posicao <= quantidade) {

        for (int i = quantidade; i > posicao; i--) {
            lista[i] = lista[i - 1];
        }

        lista[posicao] = valor;
        quantidade++;

    } else {
        printf("Não foi possível inserir.\n");
    }

    printf("Lista após inserção:\n");

    for (int i = 0; i < quantidade; i++) {
        printf("%d ", lista[i]);
    }

    return 0;
}
````

## Ideia do deslocamento

O deslocamento precisa começar do final para o início.

Se fizéssemos o deslocamento do início para o final, poderíamos sobrescrever valores importantes.

Sequência do deslocamento:

```text
40 vai para a próxima posição
30 vai para a próxima posição
25 entra na posição livre
```

# Remoção em vetor

Para remover um elemento do vetor, os elementos seguintes precisam ser deslocados para a esquerda.

Exemplo: remover o elemento da posição `1`.

Antes:

```text
Índice:   0    1    2    3
Valor:   10   20   30   40
```

Depois:

```text
Índice:   0    1    2
Valor:   10   30   40
```

````python
%%writefile codigos_c/exemplo_04.c

#include <stdio.h>

#define TAM 10

int main() {
    int lista[TAM] = {10, 20, 30, 40};
    int quantidade = 4;
    int posicao = 1;

    if (posicao >= 0 && posicao < quantidade) {

        for (int i = posicao; i < quantidade - 1; i++) {
            lista[i] = lista[i + 1];
        }

        quantidade--;

    } else {
        printf("Posição inválida.\n");
    }

    printf("Lista após remoção:\n");

    for (int i = 0; i < quantidade; i++) {
        printf("%d ", lista[i]);
    }

    return 0;
}
````

# Busca em vetor

A busca consiste em procurar um determinado valor dentro da lista.

Neste exemplo, vamos procurar o valor `30`.

````python
%%writefile codigos_c/exemplo_05.c

#include <stdio.h>

#define TAM 10

int main() {
    int lista[TAM] = {10, 20, 30, 40, 50};
    int quantidade = 5;
    int valorBuscado = 30;
    int encontrado = 0;

    for (int i = 0; i < quantidade; i++) {
        if (lista[i] == valorBuscado) {
            printf("Valor encontrado na posição %d.\n", i);
            encontrado = 1;
            break;
        }
    }

    if (!encontrado) {
        printf("Valor não encontrado.\n");
    }

    return 0;
}
````

# Limitações do vetor

O vetor é uma estrutura muito importante, mas possui limitações.

## Pontos positivos

```text
- Implementação simples.
- Acesso rápido por índice.
- Boa organização na memória.
- Excelente para listas de tamanho conhecido.
```

## Pontos negativos

```text
- Tamanho geralmente fixo.
- Pode desperdiçar memória.
- Inserção no início ou no meio exige deslocamento.
- Remoção no início ou no meio exige deslocamento.
```

## Exemplo prático

Se uma turma sempre possui no máximo 40 alunos, um vetor pode ser adequado.

Mas se a quantidade de elementos cresce e diminui constantemente, uma lista encadeada pode ser mais interessante.

# Lista encadeada simples

## Conceito

Uma **lista encadeada simples** é formada por nós.

Cada nó possui duas partes:

```text
1. O valor armazenado
2. O endereço do próximo nó
```

Representação:

```text
[10 | próximo] -> [20 | próximo] -> [30 | NULL]
```

O último nó aponta para `NULL`, indicando o fim da lista.

## Diferença em relação ao vetor

No vetor, os elementos ficam lado a lado na memória.

Na lista encadeada, os elementos podem estar espalhados na memória. O que mantém a sequência é o ponteiro para o próximo nó.

# Ponteiros e alocação dinâmica

Para trabalhar com lista encadeada em C, precisamos usar ponteiros e alocação dinâmica de memória.

## Ponteiro

Um ponteiro armazena o endereço de uma variável ou estrutura.

Exemplo:

```c
int x = 10;
int *p = &x;
```

## Alocação dinâmica

A função `malloc` reserva memória durante a execução do programa.

Exemplo:

```c
No *novo = malloc(sizeof(No));
```

A função `free` libera a memória reservada.

Exemplo:

```c
free(novo);
```

## Importante

Sempre que usamos `malloc`, em algum momento devemos usar `free` para evitar desperdício de memória.

# Criando manualmente uma lista encadeada simples

Neste exemplo, criaremos três nós manualmente.

````python
%%writefile codigos_c/exemplo_07.c

#include <stdio.h>
#include <stdlib.h>

typedef struct No {
    int valor;
    struct No *proximo;
} No;

int main() {
    No *primeiro = malloc(sizeof(No));
    No *segundo = malloc(sizeof(No));
    No *terceiro = malloc(sizeof(No));

    primeiro->valor = 10;
    primeiro->proximo = segundo;

    segundo->valor = 20;
    segundo->proximo = terceiro;

    terceiro->valor = 30;
    terceiro->proximo = NULL;

    No *atual = primeiro;

    while (atual != NULL) {
        printf("%d -> ", atual->valor);
        atual = atual->proximo;
    }

    printf("NULL\n");

    free(primeiro);
    free(segundo);
    free(terceiro);

    return 0;
}
````

# Inserção no início da lista encadeada simples

Inserir no início é uma operação eficiente.

Situação inicial:

```text
20 -> 30 -> NULL
```

Inserindo `10` no início:

```text
10 -> 20 -> 30 -> NULL
```

O novo nó passa a apontar para o antigo início.

````python
%%writefile codigos_c/exemplo_08.c

#include <stdio.h>
#include <stdlib.h>

typedef struct No {
    int valor;
    struct No *proximo;
} No;

No* inserirInicio(No *inicio, int valor) {
    No *novo = malloc(sizeof(No));

    if (novo == NULL) {
        printf("Erro ao alocar memória.\n");
        return inicio;
    }

    novo->valor = valor;
    novo->proximo = inicio;

    return novo;
}

void exibirLista(No *inicio) {
    No *atual = inicio;

    while (atual != NULL) {
        printf("%d -> ", atual->valor);
        atual = atual->proximo;
    }

    printf("NULL\n");
}

void liberarLista(No *inicio) {
    No *atual = inicio;

    while (atual != NULL) {
        No *proximo = atual->proximo;
        free(atual);
        atual = proximo;
    }
}

int main() {
    No *lista = NULL;

    lista = inserirInicio(lista, 30);
    lista = inserirInicio(lista, 20);
    lista = inserirInicio(lista, 10);

    exibirLista(lista);

    liberarLista(lista);

    return 0;
}
````

# Inserção no final da lista encadeada simples

Para inserir no final, precisamos percorrer a lista até encontrar o último nó.

O último nó é aquele cujo `proximo` é `NULL`.

```text
10 -> 20 -> 30 -> NULL
```

Inserindo `40`:

```text
10 -> 20 -> 30 -> 40 -> NULL
```

````python
%%writefile codigos_c/exemplo_09.c

#include <stdio.h>
#include <stdlib.h>

typedef struct No {
    int valor;
    struct No *proximo;
} No;

No* inserirFinal(No *inicio, int valor) {
    No *novo = malloc(sizeof(No));

    if (novo == NULL) {
        printf("Erro ao alocar memória.\n");
        return inicio;
    }

    novo->valor = valor;
    novo->proximo = NULL;

    if (inicio == NULL) {
        return novo;
    }

    No *atual = inicio;

    while (atual->proximo != NULL) {
        atual = atual->proximo;
    }

    atual->proximo = novo;

    return inicio;
}

void exibirLista(No *inicio) {
    No *atual = inicio;

    while (atual != NULL) {
        printf("%d -> ", atual->valor);
        atual = atual->proximo;
    }

    printf("NULL\n");
}

void liberarLista(No *inicio) {
    No *atual = inicio;

    while (atual != NULL) {
        No *proximo = atual->proximo;
        free(atual);
        atual = proximo;
    }
}

int main() {
    No *lista = NULL;

    lista = inserirFinal(lista, 10);
    lista = inserirFinal(lista, 20);
    lista = inserirFinal(lista, 30);

    exibirLista(lista);

    liberarLista(lista);

    return 0;
}
````

# Busca em lista encadeada simples

A busca percorre a lista nó por nó até encontrar o valor desejado ou chegar ao fim.

````python
%%writefile codigos_c/exemplo_10.c

#include <stdio.h>
#include <stdlib.h>
typedef struct No {
    int valor;
    struct No *proximo;
} No;

void buscarValor(No *inicio, int valor) {
    No *atual = inicio;
    int posicao = 0;

    while (atual != NULL) {
        if (atual->valor == valor) {
            printf("Valor %d encontrado na posição %d.\n", valor, posicao);
            return;
        }

        atual = atual->proximo;
        posicao++;
    }

    printf("Valor %d não encontrado.\n", valor);
}

int main() {
    return 0;
}
````

# Remoção em lista encadeada simples

Para remover um nó, precisamos ajustar os ponteiros.

Exemplo:

```text
10 -> 20 -> 30 -> NULL
```

Removendo `20`:

```text
10 -> 30 -> NULL
```

O nó `10` deixa de apontar para `20` e passa a apontar para `30`.

````python
%%writefile codigos_c/exemplo_11.c

#include <stdio.h>
#include <stdlib.h>
typedef struct No {
    int valor;
    struct No *anterior;
    struct No *proximo;
} No;

No* removerValor(No *inicio, int valor) {
    No *atual = inicio;
    No *anterior = NULL;

    while (atual != NULL && atual->valor != valor) {
        anterior = atual;
        atual = atual->proximo;
    }

    if (atual == NULL) {
        printf("Valor %d não encontrado.\n", valor);
        return inicio;
    }

    if (anterior == NULL) {
        inicio = atual->proximo;
    } else {
        anterior->proximo = atual->proximo;
    }

    free(atual);

    return inicio;
}

int main() {
    return 0;
}
````

# Exemplo completo: lista encadeada simples

O código a seguir reúne inserção, exibição, busca, remoção e liberação de memória.

````python
%%writefile codigos_c/exemplo_12.c

#include <stdio.h>
#include <stdlib.h>

typedef struct No {
    int valor;
    struct No *proximo;
} No;

No* inserirFinal(No *inicio, int valor) {
    No *novo = malloc(sizeof(No));

    if (novo == NULL) {
        printf("Erro ao alocar memória.\n");
        return inicio;
    }

    novo->valor = valor;
    novo->proximo = NULL;

    if (inicio == NULL) {
        return novo;
    }

    No *atual = inicio;

    while (atual->proximo != NULL) {
        atual = atual->proximo;
    }

    atual->proximo = novo;

    return inicio;
}

No* removerValor(No *inicio, int valor) {
    No *atual = inicio;
    No *anterior = NULL;

    while (atual != NULL && atual->valor != valor) {
        anterior = atual;
        atual = atual->proximo;
    }

    if (atual == NULL) {
        printf("Valor %d não encontrado.\n", valor);
        return inicio;
    }

    if (anterior == NULL) {
        inicio = atual->proximo;
    } else {
        anterior->proximo = atual->proximo;
    }

    free(atual);

    return inicio;
}

void buscarValor(No *inicio, int valor) {
    No *atual = inicio;
    int posicao = 0;

    while (atual != NULL) {
        if (atual->valor == valor) {
            printf("Valor %d encontrado na posição %d.\n", valor, posicao);
            return;
        }

        atual = atual->proximo;
        posicao++;
    }

    printf("Valor %d não encontrado.\n", valor);
}

void exibirLista(No *inicio) {
    No *atual = inicio;

    while (atual != NULL) {
        printf("%d -> ", atual->valor);
        atual = atual->proximo;
    }

    printf("NULL\n");
}

void liberarLista(No *inicio) {
    No *atual = inicio;

    while (atual != NULL) {
        No *proximo = atual->proximo;
        free(atual);
        atual = proximo;
    }
}

int main() {
    No *lista = NULL;

    lista = inserirFinal(lista, 10);
    lista = inserirFinal(lista, 20);
    lista = inserirFinal(lista, 30);
    lista = inserirFinal(lista, 40);

    printf("Lista inicial:\n");
    exibirLista(lista);

    buscarValor(lista, 30);

    lista = removerValor(lista, 20);

    printf("Lista após remover 20:\n");
    exibirLista(lista);

    liberarLista(lista);

    return 0;
}
````

# Lista encadeada circular

## Conceito

Na lista encadeada circular, o último nó aponta novamente para o primeiro nó.

Em uma lista simples comum:

```text
10 -> 20 -> 30 -> NULL
```

Em uma lista circular:

```text
10 -> 20 -> 30
^           |
|___________|
```

A lista circular não termina em `NULL`.

O fim lógico da lista é identificado quando voltamos ao nó inicial.

## Aplicações

Listas circulares são úteis quando os elementos precisam ser percorridos repetidamente.

Exemplos:

- escalonamento de processos em sistemas operacionais;
- alternância de jogadores em um jogo;
- playlists que retornam à primeira música;
- carrossel de imagens;
- sistemas com funcionamento cíclico.

# Inserção e exibição em lista circular simples

A estrutura do nó é a mesma da lista encadeada simples.

A diferença está no fato de que o último nó aponta para o início.

````python
%%writefile codigos_c/exemplo_13.c

#include <stdio.h>
#include <stdlib.h>

typedef struct No {
    int valor;
    struct No *proximo;
} No;

No* inserirCircular(No *inicio, int valor) {
    No *novo = malloc(sizeof(No));

    if (novo == NULL) {
        printf("Erro ao alocar memória.\n");
        return inicio;
    }

    novo->valor = valor;

    if (inicio == NULL) {
        novo->proximo = novo;
        return novo;
    }

    No *atual = inicio;

    while (atual->proximo != inicio) {
        atual = atual->proximo;
    }

    atual->proximo = novo;
    novo->proximo = inicio;

    return inicio;
}

void exibirCircular(No *inicio) {
    if (inicio == NULL) {
        printf("Lista vazia.\n");
        return;
    }

    No *atual = inicio;

    do {
        printf("%d -> ", atual->valor);
        atual = atual->proximo;
    } while (atual != inicio);

    printf("(volta ao início)\n");
}

void liberarCircular(No *inicio) {
    if (inicio == NULL) {
        return;
    }

    No *atual = inicio->proximo;

    while (atual != inicio) {
        No *proximo = atual->proximo;
        free(atual);
        atual = proximo;
    }

    free(inicio);
}

int main() {
    No *lista = NULL;

    lista = inserirCircular(lista, 10);
    lista = inserirCircular(lista, 20);
    lista = inserirCircular(lista, 30);

    exibirCircular(lista);

    liberarCircular(lista);

    return 0;
}
````

## Atenção ao percorrer uma lista circular

Em uma lista encadeada simples, paramos quando encontramos `NULL`.

Em uma lista circular, não existe `NULL` no final.

Por isso, usamos `do while` e paramos quando voltamos ao nó inicial.

```text
começa no início
visita os nós
para quando atual voltar a ser igual ao início
```

# Exibição reversa em lista duplamente encadeada

Para exibir em ordem reversa, primeiro precisamos chegar ao último nó.

Depois usamos o ponteiro `anterior`.

````python
%%writefile codigos_c/exemplo_16.c

#include <stdio.h>
#include <stdlib.h>

typedef struct No {
    int valor;
    struct No *anterior;
    struct No *proximo;
} No;

No* inserirFinal(No *inicio, int valor) {
    No *novo = malloc(sizeof(No));

    if (novo == NULL) {
        printf("Erro ao alocar memória.\n");
        return inicio;
    }

    novo->valor = valor;
    novo->proximo = NULL;

    if (inicio == NULL) {
        novo->anterior = NULL;
        return novo;
    }

    No *atual = inicio;

    while (atual->proximo != NULL) {
        atual = atual->proximo;
    }

    atual->proximo = novo;
    novo->anterior = atual;

    return inicio;
}

void exibirDireto(No *inicio) {
    No *atual = inicio;

    while (atual != NULL) {
        printf("%d -> ", atual->valor);
        atual = atual->proximo;
    }

    printf("NULL\n");
}

void exibirReverso(No *inicio) {
    if (inicio == NULL) {
        printf("Lista vazia.\n");
        return;
    }

    No *atual = inicio;

    while (atual->proximo != NULL) {
        atual = atual->proximo;
    }

    while (atual != NULL) {
        printf("%d -> ", atual->valor);
        atual = atual->anterior;
    }

    printf("NULL\n");
}

void liberarLista(No *inicio) {
    No *atual = inicio;

    while (atual != NULL) {
        No *proximo = atual->proximo;
        free(atual);
        atual = proximo;
    }
}

int main() {
    No *lista = NULL;

    lista = inserirFinal(lista, 10);
    lista = inserirFinal(lista, 20);
    lista = inserirFinal(lista, 30);

    printf("Percurso direto:\n");
    exibirDireto(lista);

    printf("Percurso reverso:\n");
    exibirReverso(lista);

    liberarLista(lista);

    return 0;
}
````

# Remoção em lista duplamente encadeada

Na lista duplamente encadeada, a remoção exige ajustar dois ponteiros:

```text
anterior
próximo
```

Exemplo:

```text
10 <-> 20 <-> 30
```

Removendo `20`:

```text
10 <-> 30
```

````python
%%writefile codigos_c/exemplo_17.c

#include <stdio.h>
#include <stdlib.h>
typedef struct No {
    int valor;
    struct No *anterior;
    struct No *proximo;
} No;

No* removerValor(No *inicio, int valor) {
    No *atual = inicio;

    while (atual != NULL && atual->valor != valor) {
        atual = atual->proximo;
    }

    if (atual == NULL) {
        printf("Valor %d não encontrado.\n", valor);
        return inicio;
    }

    if (atual->anterior != NULL) {
        atual->anterior->proximo = atual->proximo;
    } else {
        inicio = atual->proximo;
    }

    if (atual->proximo != NULL) {
        atual->proximo->anterior = atual->anterior;
    }

    free(atual);

    return inicio;
}

int main() {
    return 0;
}
````

# Lista duplamente encadeada circular

## Conceito

A lista duplamente encadeada circular combina duas ideias:

```text
1. Cada nó aponta para o próximo e para o anterior.
2. O último nó aponta para o primeiro.
3. O primeiro nó aponta de volta para o último.
```

Representação:

```text
______________________
|                    \/
10 <-> 20 <-> 30 <-> 40
/\____________________|
```

De forma lógica:

```text
primeiro->anterior = último
último->proximo = primeiro
```

## Aplicações

- playlists circulares com opção de voltar e avançar;
- carrosséis de imagens;
- jogos em turnos;
- gerenciamento circular de processos;
- menus circulares em interfaces.

# Inserção em lista duplamente encadeada circular

Neste exemplo, vamos inserir elementos no final da lista.

````python
%%writefile codigos_c/exemplo_18.c

#include <stdio.h>
#include <stdlib.h>

typedef struct No {
    int valor;
    struct No *anterior;
    struct No *proximo;
} No;

No* inserirDuplaCircular(No *inicio, int valor) {
    No *novo = malloc(sizeof(No));

    if (novo == NULL) {
        printf("Erro ao alocar memória.\n");
        return inicio;
    }

    novo->valor = valor;

    if (inicio == NULL) {
        novo->proximo = novo;
        novo->anterior = novo;
        return novo;
    }

    No *ultimo = inicio->anterior;

    novo->proximo = inicio;
    novo->anterior = ultimo;

    ultimo->proximo = novo;
    inicio->anterior = novo;

    return inicio;
}

void exibirDireto(No *inicio) {
    if (inicio == NULL) {
        printf("Lista vazia.\n");
        return;
    }

    No *atual = inicio;

    do {
        printf("%d <-> ", atual->valor);
        atual = atual->proximo;
    } while (atual != inicio);

    printf("(início)\n");
}

void exibirReverso(No *inicio) {
    if (inicio == NULL) {
        printf("Lista vazia.\n");
        return;
    }

    No *ultimo = inicio->anterior;
    No *atual = ultimo;

    do {
        printf("%d <-> ", atual->valor);
        atual = atual->anterior;
    } while (atual != ultimo);

    printf("(último)\n");
}

void liberarDuplaCircular(No *inicio) {
    if (inicio == NULL) {
        return;
    }

    No *atual = inicio->proximo;

    while (atual != inicio) {
        No *proximo = atual->proximo;
        free(atual);
        atual = proximo;
    }

    free(inicio);
}

int main() {
    No *lista = NULL;

    lista = inserirDuplaCircular(lista, 10);
    lista = inserirDuplaCircular(lista, 20);
    lista = inserirDuplaCircular(lista, 30);
    lista = inserirDuplaCircular(lista, 40);

    printf("Percurso direto:\n");
    exibirDireto(lista);

    printf("Percurso reverso:\n");
    exibirReverso(lista);

    liberarDuplaCircular(lista);

    return 0;
}
````

# Exemplos específicos de listas lineares: pilha e fila

Além das listas lineares gerais, duas estruturas aparecem com muita frequência em algoritmos, sistemas operacionais, compiladores, navegadores, aplicativos e sistemas de atendimento:

- **Pilha**;
- **Fila**.

As duas são listas lineares porque armazenam elementos em sequência. A diferença está na regra de entrada e saída dos dados.

```text
Lista linear geral:
permite inserir, remover e acessar elementos conforme a implementação.

Pilha:
o último que entra é o primeiro que sai.

Fila:
o primeiro que entra é o primeiro que sai.
```


## Pilha

A **pilha** segue a regra **LIFO**:

```text
LIFO = Last In, First Out
Último a entrar, primeiro a sair
```

Exemplo do dia a dia:

```text
Pilha de pratos

Coloca prato 1
Coloca prato 2
Coloca prato 3

Ao retirar:
Sai primeiro o prato 3
Depois o prato 2
Depois o prato 1
```

Representação:

```text
Topo
 ↓
[30]
[20]
[10]
```

Operações principais:

```text
push: inserir no topo
pop: remover do topo
peek/top: consultar o topo
isEmpty: verificar se está vazia
```


## Pilha usando vetor em C

Na pilha com vetor, usamos uma variável chamada `topo` para indicar a última posição ocupada.

Quando a pilha está vazia:

```text
topo = -1
```

Quando inserimos um elemento:

```text
topo aumenta uma posição
valor é armazenado em pilha[topo]
```

Quando removemos um elemento:

```text
valor do topo é removido
topo diminui uma posição
```


````python
%%writefile codigos_c/exemplo_19.c

#include <stdio.h>
#define TAM 5

void push(int pilha[], int *topo, int valor) {
    if (*topo == TAM - 1) {
        printf("Pilha cheia. Não é possível inserir %d.\n", valor);
        return;
    }

    (*topo)++;
    pilha[*topo] = valor;
    printf("Valor %d inserido na pilha.\n", valor);
}

int pop(int pilha[], int *topo) {
    if (*topo == -1) {
        printf("Pilha vazia. Não é possível remover.\n");
        return -1;
    }

    int valor = pilha[*topo];
    (*topo)--;
    return valor;
}

int peek(int pilha[], int topo) {
    if (topo == -1) {
        printf("Pilha vazia.\n");
        return -1;
    }

    return pilha[topo];
}

void exibirPilha(int pilha[], int topo) {
    if (topo == -1) {
        printf("Pilha vazia.\n");
        return;
    }

    printf("Pilha do topo para a base:\n");

    for (int i = topo; i >= 0; i--) {
        printf("[%d]\n", pilha[i]);
    }
}

int main() {
    int pilha[TAM];
    int topo = -1;

    push(pilha, &topo, 10);
    push(pilha, &topo, 20);
    push(pilha, &topo, 30);

    exibirPilha(pilha, topo);

    printf("Topo atual: %d\n", peek(pilha, topo));

    printf("Removido: %d\n", pop(pilha, &topo));

    exibirPilha(pilha, topo);

    return 0;
}
````

## Pilha usando lista encadeada em C

Na pilha com lista encadeada, o topo será o primeiro nó da lista.

Inserir no topo significa inserir no início da lista.

Remover do topo significa remover o primeiro nó da lista.

Representação:

```text
Topo
 ↓
[30] -> [20] -> [10] -> NULL
```

Essa implementação não depende de um tamanho fixo, pois os nós são criados dinamicamente com `malloc`.


````python
%%writefile codigos_c/exemplo_20.c

#include <stdio.h>
#include <stdlib.h>

typedef struct No {
    int valor;
    struct No *proximo;
} No;

No* push(No *topo, int valor) {
    No *novo = malloc(sizeof(No));

    if (novo == NULL) {
        printf("Erro ao alocar memória.\n");
        return topo;
    }

    novo->valor = valor;
    novo->proximo = topo;

    printf("Valor %d inserido na pilha.\n", valor);

    return novo;
}

No* pop(No *topo) {
    if (topo == NULL) {
        printf("Pilha vazia. Não é possível remover.\n");
        return NULL;
    }

    No *remover = topo;
    int valor = remover->valor;

    topo = topo->proximo;

    free(remover);

    printf("Valor %d removido da pilha.\n", valor);

    return topo;
}

void exibirPilha(No *topo) {
    if (topo == NULL) {
        printf("Pilha vazia.\n");
        return;
    }

    No *atual = topo;

    printf("Topo -> ");

    while (atual != NULL) {
        printf("%d -> ", atual->valor);
        atual = atual->proximo;
    }

    printf("NULL\n");
}

void liberarPilha(No *topo) {
    No *atual = topo;

    while (atual != NULL) {
        No *proximo = atual->proximo;
        free(atual);
        atual = proximo;
    }
}

int main() {
    No *topo = NULL;

    topo = push(topo, 10);
    topo = push(topo, 20);
    topo = push(topo, 30);

    exibirPilha(topo);

    topo = pop(topo);

    exibirPilha(topo);

    liberarPilha(topo);

    return 0;
}
````

# Fila

A **fila** segue a regra **FIFO**:

```text
FIFO = First In, First Out
Primeiro a entrar, primeiro a sair
```

Exemplo do dia a dia:

```text
Fila de banco

Pessoa 1 chega primeiro
Pessoa 2 chega depois
Pessoa 3 chega depois

Ao atender:
Pessoa 1 sai primeiro
Pessoa 2 sai depois
Pessoa 3 sai depois
```

Representação:

```text
Início                   Fim
 ↓                        ↓
[10] -> [20] -> [30] -> [40]
```

Operações principais:

```text
enqueue: inserir no fim da fila
dequeue: remover do início da fila
front: consultar o primeiro elemento
isEmpty: verificar se está vazia
```


## Fila usando vetor em C

Em uma fila com vetor, controlamos duas posições:

```text
inicio: indica quem será removido primeiro
fim: indica onde o próximo valor será inserido
```

Nesta primeira versão, a fila avança no vetor. É uma implementação didática e simples.

Limitação desta versão:

```text
Após muitas remoções, podem sobrar espaços vazios no começo do vetor.
Por isso, em aplicações reais, costuma-se usar fila circular.
```


````python
%%writefile codigos_c/exemplo_21.c

#include <stdio.h>
#define TAM 5

void enqueue(int fila[], int *fim, int valor) {
    if (*fim == TAM - 1) {
        printf("Fila cheia. Não é possível inserir %d.\n", valor);
        return;
    }

    (*fim)++;
    fila[*fim] = valor;
    printf("Valor %d inserido na fila.\n", valor);
}

int dequeue(int fila[], int *inicio, int fim) {
    if (*inicio > fim) {
        printf("Fila vazia. Não é possível remover.\n");
        return -1;
    }

    int valor = fila[*inicio];
    (*inicio)++;

    return valor;
}

void exibirFila(int fila[], int inicio, int fim) {
    if (inicio > fim) {
        printf("Fila vazia.\n");
        return;
    }

    printf("Início -> ");

    for (int i = inicio; i <= fim; i++) {
        printf("%d -> ", fila[i]);
    }

    printf("Fim\n");
}

int main() {
    int fila[TAM];
    int inicio = 0;
    int fim = -1;

    enqueue(fila, &fim, 10);
    enqueue(fila, &fim, 20);
    enqueue(fila, &fim, 30);

    exibirFila(fila, inicio, fim);

    printf("Removido: %d\n", dequeue(fila, &inicio, fim));

    exibirFila(fila, inicio, fim);

    return 0;
}
````

## Fila circular usando vetor

Para aproveitar melhor o vetor, podemos implementar uma **fila circular**.

A ideia é fazer o índice voltar para o começo quando chega ao final do vetor.

```text
Índices do vetor:
0  1  2  3  4

Quando o fim chega ao índice 4,
o próximo avanço pode voltar para o índice 0,
se houver espaço disponível.
```

A movimentação circular é feita com o operador `%`:

```c
proxima_posicao = (posicao_atual + 1) % TAM;
```

Essa técnica é muito usada em buffers, filas de impressão, comunicação entre processos e sistemas embarcados.


````python
%%writefile codigos_c/exemplo_22.c

#include <stdio.h>
#define TAM 5

typedef struct {
    int dados[TAM];
    int inicio;
    int fim;
    int quantidade;
} FilaCircular;

void inicializar(FilaCircular *fila) {
    fila->inicio = 0;
    fila->fim = -1;
    fila->quantidade = 0;
}

void enqueue(FilaCircular *fila, int valor) {
    if (fila->quantidade == TAM) {
        printf("Fila cheia. Não é possível inserir %d.\n", valor);
        return;
    }

    fila->fim = (fila->fim + 1) % TAM;
    fila->dados[fila->fim] = valor;
    fila->quantidade++;

    printf("Valor %d inserido.\n", valor);
}

int dequeue(FilaCircular *fila) {
    if (fila->quantidade == 0) {
        printf("Fila vazia. Não é possível remover.\n");
        return -1;
    }

    int valor = fila->dados[fila->inicio];
    fila->inicio = (fila->inicio + 1) % TAM;
    fila->quantidade--;

    return valor;
}

void exibirFila(FilaCircular fila) {
    if (fila.quantidade == 0) {
        printf("Fila vazia.\n");
        return;
    }

    printf("Início -> ");

    int posicao = fila.inicio;

    for (int i = 0; i < fila.quantidade; i++) {
        printf("%d -> ", fila.dados[posicao]);
        posicao = (posicao + 1) % TAM;
    }

    printf("Fim\n");
}

int main() {
    FilaCircular fila;

    inicializar(&fila);

    enqueue(&fila, 10);
    enqueue(&fila, 20);
    enqueue(&fila, 30);
    enqueue(&fila, 40);

    exibirFila(fila);

    printf("Removido: %d\n", dequeue(&fila));
    printf("Removido: %d\n", dequeue(&fila));

    enqueue(&fila, 50);
    enqueue(&fila, 60);

    exibirFila(fila);

    return 0;
}
````

## Fila usando lista encadeada em C

Na fila com lista encadeada, usamos dois ponteiros:

```text
inicio: aponta para o primeiro elemento da fila
fim: aponta para o último elemento da fila
```

Inserir na fila significa adicionar no final.

Remover da fila significa retirar do início.

Representação:

```text
inicio                          fim
  ↓                              ↓
[10] -> [20] -> [30] -> [40] -> NULL
```


````python
%%writefile codigos_c/exemplo_23.c

#include <stdio.h>
#include <stdlib.h>

typedef struct No {
    int valor;
    struct No *proximo;
} No;

typedef struct {
    No *inicio;
    No *fim;
} Fila;

void inicializar(Fila *fila) {
    fila->inicio = NULL;
    fila->fim = NULL;
}

void enqueue(Fila *fila, int valor) {
    No *novo = malloc(sizeof(No));

    if (novo == NULL) {
        printf("Erro ao alocar memória.\n");
        return;
    }

    novo->valor = valor;
    novo->proximo = NULL;

    if (fila->inicio == NULL) {
        fila->inicio = novo;
        fila->fim = novo;
    } else {
        fila->fim->proximo = novo;
        fila->fim = novo;
    }

    printf("Valor %d inserido na fila.\n", valor);
}

int dequeue(Fila *fila) {
    if (fila->inicio == NULL) {
        printf("Fila vazia. Não é possível remover.\n");
        return -1;
    }

    No *remover = fila->inicio;
    int valor = remover->valor;

    fila->inicio = fila->inicio->proximo;

    if (fila->inicio == NULL) {
        fila->fim = NULL;
    }

    free(remover);

    return valor;
}

void exibirFila(Fila fila) {
    if (fila.inicio == NULL) {
        printf("Fila vazia.\n");
        return;
    }

    No *atual = fila.inicio;

    printf("Início -> ");

    while (atual != NULL) {
        printf("%d -> ", atual->valor);
        atual = atual->proximo;
    }

    printf("Fim\n");
}

void liberarFila(Fila *fila) {
    while (fila->inicio != NULL) {
        dequeue(fila);
    }
}

int main() {
    Fila fila;

    inicializar(&fila);

    enqueue(&fila, 10);
    enqueue(&fila, 20);
    enqueue(&fila, 30);

    exibirFila(fila);

    printf("Removido: %d\n", dequeue(&fila));

    exibirFila(fila);

    liberarFila(&fila);

    return 0;
}
````
