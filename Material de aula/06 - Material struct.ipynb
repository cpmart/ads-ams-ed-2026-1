## 1. O que ĂŠ `struct`?

Em C, `struct` ĂŠ uma **estrutura de dados composta** que permite **agrupar diferentes tipos de variĂĄveis dentro de um mesmo bloco**.

Ela ĂŠ usada quando precisamos representar uma entidade do mundo real com vĂĄrias caracterĂ­sticas.

### Exemplo de ideia
Um aluno pode ter:
- nome
- idade
- nota

Esses dados sĂŁo diferentes entre si, mas pertencem ao mesmo objeto: **o aluno**.

Com `struct`, podemos organizar isso melhor.


## 2. DefiniĂ§ĂŁo bĂĄsica

A sintaxe geral ĂŠ:


```python
struct NomeDaStruct {
    tipo campo1;
    tipo campo2;
    tipo campo3;
};
```

### Exemplo


```python
struct Aluno {
    char nome[50];
    int idade;
    float nota;
};
```

Nesse exemplo, foi criada uma estrutura chamada `Aluno` com trĂŞs campos:
- `nome`
- `idade`
- `nota`


## 3. Como declarar variĂĄveis do tipo `struct`

Depois de definir a estrutura, podemos criar variĂĄveis desse tipo:


```python
struct Aluno {
    char nome[50];
    int idade;
    float nota;
};

int main() {
    struct Aluno a1;
    return 0;
}
```

Aqui, `a1` ĂŠ uma variĂĄvel do tipo `struct Aluno`.


## 4. Como acessar os campos

Usamos o **operador ponto (`.`)** para acessar os campos da estrutura.


```python
#include <stdio.h>
#include <string.h>

struct Aluno {
    char nome[50];
    int idade;
    float nota;
};

int main() {
    struct Aluno a1;

    strcpy(a1.nome, "Carlos");
    a1.idade = 20;
    a1.nota = 8.5;

    printf("Nome: %s\n", a1.nome);
    printf("Idade: %d\n", a1.idade);
    printf("Nota: %.2f\n", a1.nota);

    return 0;
}
```

### SaĂ­da esperada


```python
Nome: Carlos
Idade: 20
Nota: 8.50
```

## 5. Por que usar `struct`?

Sem `struct`, terĂ­amos que criar vĂĄrias variĂĄveis separadas:


```python
char nome[50];
int idade;
float nota;
```

Isso funciona para um Ăşnico aluno, mas fica ruim quando hĂĄ muitos registros.

Com `struct`, conseguimos organizar melhor os dados.

### Vantagens
- melhora a organizaĂ§ĂŁo do cĂłdigo
- agrupa dados relacionados
- facilita a leitura e manutenĂ§ĂŁo
- permite modelar objetos do mundo real
- ajuda na criaĂ§ĂŁo de vetores de registros

## 7. Exemplo prĂĄtico: cadastro de produto


```python
#include <stdio.h>
#include <string.h>

struct Produto {
    int codigo;
    char nome[30];
    float preco;
};

int main() {
    struct Produto p1;

    p1.codigo = 101;
    strcpy(p1.nome, "Teclado");
    p1.preco = 150.75;

    printf("Codigo: %d\n", p1.codigo);
    printf("Nome: %s\n", p1.nome);
    printf("Preco: R$ %.2f\n", p1.preco);

    return 0;
}
```

## 8. Lendo dados com `scanf`

TambĂŠm podemos preencher a estrutura com dados digitados pelo usuĂĄrio.


```python
#include <stdio.h>

struct Pessoa {
    char nome[50];
    int idade;
};

int main() {
    struct Pessoa p;

    printf("Digite o nome: ");
    scanf("%49s", p.nome);

    printf("Digite a idade: ");
    scanf("%d", &p.idade);

    printf("\nDados informados:\n");
    printf("Nome: %s\n", p.nome);
    printf("Idade: %d\n", p.idade);

    return 0;
}
```

### ObservaĂ§ĂŁo
`scanf("%s", ...)` lĂŞ apenas uma palavra. Para nomes completos com espaĂ§o, normalmente usamos `fgets`.


## 9. Exemplo com vetor de `struct`

Uma das maiores utilidades de `struct` ĂŠ criar vetores de registros.


```python
#include <stdio.h>
#include <string.h>

struct Aluno {
    char nome[50];
    float nota;
};

int main() {
    struct Aluno turma[3];

    strcpy(turma[0].nome, "Ana");
    turma[0].nota = 9.0;

    strcpy(turma[1].nome, "Bruno");
    turma[1].nota = 7.5;

    strcpy(turma[2].nome, "Clara");
    turma[2].nota = 8.8;

    for (int i = 0; i < 3; i++) {
        printf("Aluno: %s | Nota: %.2f\n", turma[i].nome, turma[i].nota);
    }

    return 0;
}
```

### Ideia importante
Aqui temos um vetor em que cada posiĂ§ĂŁo guarda um registro completo.


## 10. `typedef` com `struct`

Em C, ĂŠ comum usar `typedef` para simplificar a declaraĂ§ĂŁo.

### Sem `typedef`


```python
struct Aluno {
    char nome[50];
    int idade;
};
```

Para declarar:


```python
struct Aluno a1;
```

### Com `typedef`


```python
typedef struct {
    char nome[50];
    int idade;
} Aluno;
```

Agora basta fazer:


```python
Aluno a1;
```

### Exemplo completo


```python
#include <stdio.h>
#include <string.h>

typedef struct {
    char nome[50];
    int idade;
} Aluno;

int main() {
    Aluno a1;

    strcpy(a1.nome, "Mariana");
    a1.idade = 22;

    printf("Nome: %s\n", a1.nome);
    printf("Idade: %d\n", a1.idade);

    return 0;
}
```

## 11. `struct` dentro de outra `struct`

TambĂŠm podemos criar estruturas compostas por outras estruturas.


```python
#include <stdio.h>

typedef struct {
    int dia;
    int mes;
    int ano;
} Data;

typedef struct {
    char nome[50];
    Data nascimento;
} Pessoa;

int main() {
    Pessoa p = {"Juliana", {15, 8, 2000}};

    printf("Nome: %s\n", p.nome);
    printf("Nascimento: %02d/%02d/%d\n",
           p.nascimento.dia,
           p.nascimento.mes,
           p.nascimento.ano);

    return 0;
}
```

## 12. Passando `struct` para funĂ§Ăľes

Podemos passar estruturas para funĂ§Ăľes.


```python
#include <stdio.h>
#include <string.h>

typedef struct {
    char nome[50];
    float preco;
} Produto;

void exibirProduto(Produto p) {
    printf("Nome: %s\n", p.nome);
    printf("Preco: R$ %.2f\n", p.preco);
}

int main() {
    Produto p1;

    strcpy(p1.nome, "Mouse");
    p1.preco = 89.90;

    exibirProduto(p1);

    return 0;
}
```

## 16. Exemplo mais completo


```python
#include <stdio.h>
#include <string.h>

typedef struct {
    int ra;
    char nome[50];
    float nota1;
    float nota2;
} Aluno;

int main() {
    Aluno a;

    a.ra = 123;
    strcpy(a.nome, "Fernando");
    a.nota1 = 7.5;
    a.nota2 = 8.0;

    float media = (a.nota1 + a.nota2) / 2;

    printf("RA: %d\n", a.ra);
    printf("Nome: %s\n", a.nome);
    printf("Nota 1: %.2f\n", a.nota1);
    printf("Nota 2: %.2f\n", a.nota2);
    printf("Media: %.2f\n", media);

    return 0;
}
```
