# Definição

Os **laços de repetição** são estruturas usadas para executar um mesmo bloco de código várias vezes, de forma controlada.

Eles são muito úteis quando queremos repetir uma ação sem precisar escrever o mesmo comando várias vezes.

Em C, os principais laços de repetição são:

- `while`
- `do...while`
- `for`


# Para que servem?

Os laços de repetição são usados quando precisamos:

- contar valores
- repetir cálculos
- percorrer vetores e matrizes
- validar entradas do usuário
- criar menus interativos
- executar processos até que uma condição seja satisfeita

## Exemplos práticos

- imprimir números de 1 a 10
- calcular a soma de vários valores
- pedir senha até que esteja correta
- mostrar um menu até o usuário escolher sair


# Estrutura `while`

O `while` executa um bloco de código **enquanto** uma condição for verdadeira.

## Sintaxe

```c
while (condicao) {
    // comandos
}
```

## Funcionamento

1. testa a condição  
2. se for verdadeira, executa o bloco  
3. volta e testa novamente  
4. repete até a condição se tornar falsa  


## Exemplo: imprimir de 1 a 5

````python
#include <stdio.h>

int main() {
    int i = 1;

    while (i <= 5) {
        printf("%d\n", i);
        i++;
    }

    return 0;
}
````

## Saída esperada

```c
1
2
3
4
5
```

## Explicação

- `i` começa com 1  
- enquanto `i <= 5`, o número é exibido  
- depois `i` é incrementado com `i++`  


# Estrutura `do...while`

O `do...while` é parecido com o `while`, mas possui uma diferença importante:

**o bloco é executado pelo menos uma vez**, mesmo que a condição seja falsa.

## Sintaxe

```c
do {
    // comandos
} while (condicao);
```


## Exemplo: imprimir de 1 a 5

````python
#include <stdio.h>

int main() {
    int i = 1;

    do {
        printf("%d\n", i);
        i++;
    } while (i <= 5);

    return 0;
}
````

## Explicação

Nesse caso:

- o código executa primeiro  
- depois verifica a condição  

## Exemplo de uso comum

É muito usado em:

- menus
- validação de entrada
- programas que precisam executar pelo menos uma vez


# Estrutura `for`

O `for` é geralmente usado quando sabemos quantas vezes a repetição deve acontecer.

## Sintaxe

```c
for (inicializacao; condicao; incremento) {
    // comandos
}
```


## Exemplo: imprimir de 1 a 5

````python
#include <stdio.h>

int main() {
    int i;

    for (i = 1; i <= 5; i++) {
        printf("%d\n", i);
    }

    return 0;
}
````

## Explicação

No `for`:

- `i = 1` → inicializa a variável  
- `i <= 5` → define a condição  
- `i++` → atualiza a variável a cada repetição  


# Comparação entre `while`, `do...while` e `for`

## `while`
Use quando a repetição depende de uma condição e não se sabe exatamente quantas vezes vai repetir.

## `do...while`
Use quando o bloco precisa ser executado pelo menos uma vez.

## `for`
Use quando a quantidade de repetições é conhecida ou mais fácil de controlar com contador.


# Exemplos práticos

## Exemplo: soma dos números de 1 a 10 com `for`


````python
#include <stdio.h>

int main() {
    int i, soma = 0;

    for (i = 1; i <= 10; i++) {
        soma = soma + i;
    }

    printf("Soma = %d\n", soma);

    return 0;
}
````

## Saída esperada

```c
Soma = 55
```


## Exemplo: tabuada com `while`

````python
#include <stdio.h>

int main() {
    int numero, i = 1;

    printf("Digite um numero: ");
    scanf("%d", &numero);

    while (i <= 10) {
        printf("%d x %d = %d\n", numero, i, numero * i);
        i++;
    }

    return 0;
}
````

## Exemplo: menu simples com `do...while`

````python
#include <stdio.h>

int main() {
    int opcao;

    do {
        printf("\nMenu\n");
        printf("1 - Cadastrar\n");
        printf("2 - Consultar\n");
        printf("0 - Sair\n");
        printf("Escolha: ");
        scanf("%d", &opcao);

        if (opcao == 1) {
            printf("Opcao cadastrar selecionada.\n");
        } else if (opcao == 2) {
            printf("Opcao consultar selecionada.\n");
        } else if (opcao == 0) {
            printf("Encerrando o programa.\n");
        } else {
            printf("Opcao invalida.\n");
        }

    } while (opcao != 0);

    return 0;
}
````

# Laços infinitos

Um **laço infinito** acontece quando a condição nunca se torna falsa.

## Exemplo incorreto


````python
#include <stdio.h>

int main() {
    int i = 1;

    while (i <= 5) {
        printf("%d\n", i);
    }

    return 0;
}
````

## Problema

A variável `i` nunca é incrementada, então a condição `i <= 5` será sempre verdadeira.

## Correção

```c
i++;
```


# Comandos úteis em laços

## `break`
Interrompe o laço imediatamente.

### Exemplo


````python
#include <stdio.h>

int main() {
    int i;

    for (i = 1; i <= 10; i++) {
        if (i == 6) {
            break;
        }
        printf("%d\n", i);
    }

    return 0;
}
````

### Saída esperada

```c
1
2
3
4
5
```


## `continue`
Pula a repetição atual e vai para a próxima.

### Exemplo


````python
#include <stdio.h>

int main() {
    int i;

    for (i = 1; i <= 5; i++) {
        if (i == 3) {
            continue;
        }
        printf("%d\n", i);
    }

    return 0;
}
````

### Saída esperada

```c
1
2
4
5
```


# Laços aninhados

São laços dentro de outros laços.

Muito usados em:

- matrizes
- tabelas
- desenhos com caracteres
- combinações de valores

## Exemplo: imprimir uma matriz 3x3 de asteriscos


````python
#include <stdio.h>

int main() {
    int i, j;

    for (i = 1; i <= 3; i++) {
        for (j = 1; j <= 3; j++) {
            printf("* ");
        }
        printf("\n");
    }

    return 0;
}
````

## Saída esperada

```c
* * *
* * *
* * *
```

# Boas práticas

- inicialize corretamente a variável de controle
- garanta que a condição poderá se tornar falsa
- evite laços infinitos sem necessidade
- use `for` quando houver contador
- use `while` quando depender de condição
- use `do...while` quando precisar executar ao menos uma vez
- mantenha o código bem indentado

# Resumo final

Os laços de repetição em C permitem executar comandos várias vezes de forma organizada e eficiente.
