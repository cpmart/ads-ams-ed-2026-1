# Ponteiros em Linguagem C

Material didático sobre definição, aplicações, exemplos e cuidados no uso de ponteiros na linguagem C.


## 1. O que é um ponteiro?

Um ponteiro é uma variável que armazena o endereço de memória de outra variável.

Em vez de guardar diretamente um valor, como `10` ou `'A'`, o ponteiro guarda onde esse valor está armazenado na memória.

### Exemplo conceitual

Imagine a variável:

```c
int x = 10;
```

A variável `x` possui:
- um valor: `10`
- um endereço de memória: algo como `0x7ffe...`

Um ponteiro pode armazenar esse endereço:

```c
int *p;
p = &x;
```

Nesse caso:
- `p` é um ponteiro para inteiro
- `&x` significa endereço de `x`
- `p` passa a apontar para `x`


## 2. Sintaxe básica

### Declaração de ponteiro

```c
int *p;
```

Isso significa:
- `p` é uma variável
- ela pode guardar o endereço de uma variável do tipo `int`

### Operadores importantes

#### `&` - operador de endereço
Retorna o endereço de memória de uma variável.

```c
&x
```

#### `*` - operador de indireção (desreferência)
Acessa o valor armazenado no endereço apontado.

```c
*p
```


## 3. Primeiro exemplo simples

````python
#include <stdio.h>

int main() {
    int x = 25;
    int *p;

    p = &x;

    printf("Valor de x: %d\n", x);
    printf("Endereco de x: %p\n", &x);
    printf("Valor guardado em p: %p\n", p);
    printf("Valor apontado por p: %d\n", *p);

    return 0;
}
````

### Explicação

- `x` guarda `25`
- `p = &x` faz `p` guardar o endereço de `x`
- `*p` acessa o conteúdo que está nesse endereço, ou seja, `25`


## 4. Representação visual

```c
int x = 25;
int *p = &x;
```

Podemos imaginar assim:

```text
x = 25         endereço: 1000
p = 1000
*p = 25
```

Ou seja:
- `p` guarda o endereço `1000`
- nesse endereço está o valor `25`


## 5. Alterando valores com ponteiros

Uma das grandes vantagens dos ponteiros é que eles permitem alterar diretamente o valor da variável apontada.


````python
#include <stdio.h>

int main() {
    int x = 10;
    int *p = &x;

    printf("Antes: x = %d\n", x);

    *p = 50;

    printf("Depois: x = %d\n", x);

    return 0;
}
````

### Saída esperada

```text
Antes: x = 10
Depois: x = 50
```

### Explicação

Quando fazemos:

```c
*p = 50;
```

estamos alterando o valor armazenado no endereço apontado por `p`, que é o endereço de `x`.


## 6. Ponteiro e tipo de dado

O tipo do ponteiro deve ser compatível com o tipo da variável apontada.

### Exemplos

```c
int *p1;
float *p2;
char *p3;
```

### Uso

```c
int a = 5;
float b = 3.14;
char c = 'Z';

p1 = &a;
p2 = &b;
p3 = &c;
```

Isso é importante porque o compilador precisa saber:
- quantos bytes ler
- como interpretar o valor na memória


## 7. Aplicações dos ponteiros

Ponteiros são fundamentais em C e aparecem em várias situações.


### 7.1 Passagem de parâmetros por referência

Em C, normalmente os argumentos são passados por valor. Isso significa que a função recebe uma cópia.

Com ponteiros, podemos permitir que a função altere a variável original.


#### Exemplo: trocar dois valores

````python
#include <stdio.h>

void trocar(int *a, int *b) {
    int temp = *a;
    *a = *b;
    *b = temp;
}

int main() {
    int x = 10, y = 20;

    printf("Antes: x = %d, y = %d\n", x, y);

    trocar(&x, &y);

    printf("Depois: x = %d, y = %d\n", x, y);

    return 0;
}
````

### Explicação

- `trocar(&x, &y)` envia os endereços de `x` e `y`
- a função manipula diretamente os valores originais


### 7.2 Vetores e ponteiros

O nome de um vetor já representa o endereço do primeiro elemento.

```c
int v[3] = {10, 20, 30};
```

Podemos acessar o primeiro elemento assim:

```c
v[0]
```

Ou assim:

```c
*(v + 0)
```


#### Exemplo

````python
#include <stdio.h>

int main() {
    int v[3] = {10, 20, 30};

    printf("%d\n", v[0]);
    printf("%d\n", *(v + 1));
    printf("%d\n", *(v + 2));

    return 0;
}
````

### Relação importante

```c
v[i] == *(v + i)
```


### 7.3 Alocação dinâmica de memória

Ponteiros são usados para reservar memória em tempo de execução com funções como:
- `malloc()`
- `calloc()`
- `realloc()`
- `free()`


#### Exemplo com `malloc`

````python
#include <stdio.h>
#include <stdlib.h>

int main() {
    int *p;

    p = (int *) malloc(sizeof(int));

    if (p == NULL) {
        printf("Erro ao alocar memoria.\n");
        return 1;
    }

    *p = 100;

    printf("Valor armazenado: %d\n", *p);

    free(p);

    return 0;
}
````

### Explicação

- `malloc(sizeof(int))` reserva espaço para um inteiro
- `p` aponta para essa área
- `free(p)` libera a memória


### 7.4 Ponteiros em funções com vetores

Quando passamos um vetor para uma função, na prática enviamos um ponteiro para seu primeiro elemento.


#### Exemplo

````python
#include <stdio.h>

void imprimir(int *v, int tamanho) {
    for (int i = 0; i < tamanho; i++) {
        printf("%d ", v[i]);
    }
    printf("\n");
}

int main() {
    int numeros[5] = {1, 2, 3, 4, 5};

    imprimir(numeros, 5);

    return 0;
}
````

### 7.5 Estruturas de dados dinâmicas

Ponteiros são essenciais para criar:
- listas encadeadas
- pilhas
- filas
- árvores
- grafos

#### Exemplo simples de `struct` com ponteiro

```c
struct No {
    int valor;
    struct No *proximo;
};
```

Aqui, `proximo` guarda o endereço do próximo nó.


## 8. Ponteiro nulo

Um ponteiro pode ser inicializado com `NULL` quando ainda não aponta para nenhum lugar válido.

```c
int *p = NULL;
```

Isso é importante para:
- evitar lixo de memória
- indicar sem referência
- facilitar testes de segurança

### Exemplo

```c
if (p == NULL) {
    printf("Ponteiro nao inicializado.\n");
}
```


## 9. Cuidados importantes com ponteiros

### 9.1 Ponteiro não inicializado

Errado:

```c
int *p;
*p = 10;
```

Problema:
- `p` não aponta para um endereço válido
- isso pode causar erro de execução


### 9.2 Acessar memória já liberada

```c
int *p = malloc(sizeof(int));
free(p);
*p = 5;
```

Depois de `free(p)`, o ponteiro não deve mais ser usado dessa forma.

Uma boa prática é:

```c
free(p);
p = NULL;
```


### 9.3 Confundir endereço com valor

```c
printf("%p\n", p);   // mostra endereço
printf("%d\n", *p);  // mostra valor apontado
```


## 10. Exemplo completo e didático

````python
#include <stdio.h>

int main() {
    int idade = 18;
    int *ponteiro;

    ponteiro = &idade;

    printf("Idade: %d\n", idade);
    printf("Endereco de idade: %p\n", &idade);
    printf("Valor de ponteiro: %p\n", ponteiro);
    printf("Valor apontado por ponteiro: %d\n", *ponteiro);

    *ponteiro = 25;

    printf("Nova idade: %d\n", idade);

    return 0;
}
````

### O que esse programa mostra?

- como declarar um ponteiro
- como armazenar o endereço de uma variável
- como acessar o valor apontado
- como alterar a variável original por meio do ponteiro

## 12. Quando usar ponteiros?

Ponteiros são muito usados quando precisamos:
- alterar variáveis dentro de funções
- trabalhar com vetores e matrizes
- usar alocação dinâmica de memória
- construir estruturas de dados dinâmicas
- manipular memória de forma mais eficiente


## 13. Vantagens do uso de ponteiros

- permitem acesso direto à memória
- tornam possível a alocação dinâmica
- facilitam manipulação de grandes estruturas sem cópias desnecessárias
- são essenciais para estruturas dinâmicas
- permitem maior controle e eficiência


## 14. Desvantagens e dificuldades

- são mais difíceis para iniciantes
- podem causar erros graves se usados incorretamente
- exigem cuidado com inicialização e liberação de memória
- podem gerar falhas como:
  - acesso inválido
  - vazamento de memória
  - ponteiro pendente


## 15. Exemplos práticos de uso no dia a dia da programação

### Exemplo 1: função que retorna mais de um resultado

Podemos usar ponteiros para preencher variáveis fora da função.


````python
#include <stdio.h>

void calcular(int a, int b, int *soma, int *produto) {
    *soma = a + b;
    *produto = a * b;
}

int main() {
    int s, p;

    calcular(4, 5, &s, &p);

    printf("Soma = %d\n", s);
    printf("Produto = %d\n", p);

    return 0;
}
````

### Exemplo 2: percorrendo vetor com ponteiro

````python
#include <stdio.h>

int main() {
    int v[4] = {2, 4, 6, 8};
    int *p = v;

    for (int i = 0; i < 4; i++) {
        printf("%d ", *(p + i));
    }

    return 0;
}
````
